# 📘 Lesson 20: Unit Testing with `unittest`

Unit testing means checking small pieces of code, usually functions, to make sure they work correctly. Python includes a built-in testing module called `unittest`.

| Concept | Description |
|---------|-------------|
| Unit test | A small test for one function or behavior |
| Test case | A group of related tests |
| Assertion | A check that something is true |
| `unittest` | Python's built-in testing framework |
| Passing test | A test that succeeds |
| Failing test | A test that shows a problem |

---

## 🔹 1. Why Write Tests?

Tests help you catch bugs early. They also give you confidence when changing code later.

Good tests answer: **Does this function return the result I expect?**

```python
# ▶️ A simple function to test
def add(a, b):
    return a + b

print(add(2, 3))
print(add(-1, 1))
```

---

## 🔹 2. Your First `unittest` Test

A test class inherits from `unittest.TestCase`. Test method names should start with `test_`.

```python
# ▶️ Basic unittest example
import unittest

def add(a, b):
    return a + b

class TestAdd(unittest.TestCase):
    def test_add_positive_numbers(self):
        self.assertEqual(add(2, 3), 5)

    def test_add_negative_numbers(self):
        self.assertEqual(add(-2, -3), -5)

unittest.main(argv=[""], exit=False)
```

---

## 🔹 3. Common Assertions

Assertions are checks inside tests. If the assertion is false, the test fails.

| Assertion | Meaning |
|-----------|---------|
| `assertEqual(a, b)` | `a == b` |
| `assertTrue(x)` | `x` is true |
| `assertFalse(x)` | `x` is false |
| `assertIn(a, b)` | `a` is inside `b` |
| `assertRaises(Error)` | an error is raised |

```python
# ▶️ Common assertions
import unittest

def is_even(number):
    return number % 2 == 0

class TestHelpers(unittest.TestCase):
    def test_is_even_true(self):
        self.assertTrue(is_even(10))

    def test_is_even_false(self):
        self.assertFalse(is_even(7))

    def test_item_in_list(self):
        self.assertIn("apple", ["apple", "banana"])

unittest.main(argv=[""], exit=False)
```

---

## 🔹 4. Testing Edge Cases

An edge case is an unusual or boundary input, such as `0`, an empty list, or a negative number.

```python
# ▶️ Testing edge cases
import unittest

def average(numbers):
    if len(numbers) == 0:
        return 0
    return sum(numbers) / len(numbers)

class TestAverage(unittest.TestCase):
    def test_average_numbers(self):
        self.assertEqual(average([10, 20, 30]), 20)

    def test_average_empty_list(self):
        self.assertEqual(average([]), 0)

unittest.main(argv=[""], exit=False)
```

---

## 🔹 5. Testing Errors with `assertRaises`

Sometimes the correct behavior is to raise an error. Use `assertRaises` to test that.

```python
# ▶️ Testing raised errors
import unittest

def divide(a, b):
    if b == 0:
        raise ValueError("Cannot divide by zero")
    return a / b

class TestDivide(unittest.TestCase):
    def test_divide_valid(self):
        self.assertEqual(divide(10, 2), 5)

    def test_divide_by_zero(self):
        with self.assertRaises(ValueError):
            divide(10, 0)

unittest.main(argv=[""], exit=False)
```

```python
# ✅ Solution - Exercise 1: Test a Multiply Function
import unittest

def multiply(a, b):
    return a * b

class TestMultiply(unittest.TestCase):
    def test_positive_numbers(self):
        self.assertEqual(multiply(3, 4), 12)

    def test_negative_number(self):
        self.assertEqual(multiply(-2, 5), -10)

unittest.main(argv=[""], exit=False)
```

```python
# ✅ Solution - Exercise 2: Test a String Helper
import unittest

def clean_name(name):
    return name.strip().title()

class TestCleanName(unittest.TestCase):
    def test_spaces_removed(self):
        self.assertEqual(clean_name("  alice  "), "Alice")

    def test_title_case(self):
        self.assertEqual(clean_name("bob smith"), "Bob Smith")

unittest.main(argv=[""], exit=False)
```

```python
# ✅ Solution - Exercise 3: Test an Error
import unittest

def get_first_item(items):
    if len(items) == 0:
        raise ValueError("List is empty")
    return items[0]

class TestGetFirstItem(unittest.TestCase):
    def test_normal_list(self):
        self.assertEqual(get_first_item(["apple", "banana"]), "apple")

    def test_empty_list(self):
        with self.assertRaises(ValueError):
            get_first_item([])

unittest.main(argv=[""], exit=False)
```
