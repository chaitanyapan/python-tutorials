# 📘 Lesson 19: Debugging Basics

Debugging means finding and fixing problems in your code. Every programmer debugs — it is a normal part of coding.

| Concept | Description |
|---------|-------------|
| Bug | A mistake or problem in code |
| Error message | Python's clue about what went wrong |
| Traceback | The list of lines that led to an error |
| Print debugging | Using `print()` to inspect values |
| Assertion | A quick check that something should be true |

## 🔹 2. Common Bug Types

Some bugs crash your program. Others run without crashing but give the wrong answer.

| Bug Type | Example |
|----------|---------|
| Syntax error | Missing `:` or closing parenthesis |
| Runtime error | Dividing by zero |
| Logic error | Code runs but result is wrong |
| Type error | Using a value in the wrong way |

```python
# ▶️ Logic error example
price = 100
discount_percent = 20

# Bug: this subtracts 20 directly, but maybe we meant 20%
final_price = price - discount_percent
print("Final price:", final_price)

# Correct version
correct_price = price - (price * discount_percent / 100)
print("Correct price:", correct_price)
```

---

## 🔹 3. Print Debugging

A simple way to debug is to print values while the program runs.

Print debugging helps answer: **What value does this variable have right now?**

```python
# ▶️ Print debugging
def calculate_total(prices):
    total = 0
    for price in prices:
        print("Adding price:", price)
        total += price
        print("Current total:", total)
    return total

result = calculate_total([10, 5, 8])
print("Final total:", result)
```

---

## 🔹 4. Checking Types with `type()`

Sometimes a bug happens because a value is not the type you expected. Use `type()` to check.

```python
# ▶️ Inspect types
age = "25"
print(age)
print(type(age))

age = int(age)
print(age)
print(type(age))
```

---

## 🔹 5. Using `repr()` to See Hidden Details

`repr()` shows a more exact version of a value. It is useful for spotting extra spaces or newline characters.

```python
# ▶️ repr() example
username = "admin "

print(username)
print(repr(username))  # Notice the extra space before the closing quote

clean_username = username.strip()
print(repr(clean_username))
```

---

## 🔹 6. Basic Assertions

An `assert` checks that something is true. If it is false, Python raises an `AssertionError`.

> 💡 Assertions are useful for quick checks while learning and debugging.

```python
# ▶️ Basic assert
def add(a, b):
    return a + b

assert add(2, 3) == 5
assert add(-1, 1) == 0

print("All checks passed!")
```
