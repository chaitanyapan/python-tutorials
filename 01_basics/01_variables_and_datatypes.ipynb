# 📘 Lesson 1: Variables & Data Types

In Python, a **variable** is a container for storing data values.
Unlike other languages, Python has **no command for declaring a variable** — it is created the moment you assign a value to it.

---

## 🔹 1. Creating Variables
```python
name = "Alice"   # String
age = 25         # Integer
height = 5.7     # Float
is_student = True  # Boolean
```
- No need to declare types — Python figures it out automatically!
- Variable names are **case-sensitive**: `age` and `Age` are different.

````python
# ▶️ Run this cell to see variables in action!
name = "Alice"
age = 25
height = 5.7
is_student = True

print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is Student:", is_student)
````

---

## 🔹 2. Data Types

| Type | Example | Description |
|------|---------|-------------|
| `str` | `"Hello"` | Text |
| `int` | `42` | Whole numbers |
| `float` | `3.14` | Decimal numbers |
| `bool` | `True / False` | Boolean values |
| `NoneType` | `None` | Represents no value |

Use `type()` to check the type of a variable.

````python
# ▶️ Check types using type()
print(type("Hello"))   # <class 'str'>
print(type(42))        # <class 'int'>
print(type(3.14))      # <class 'float'>
print(type(True))      # <class 'bool'>
print(type(None))      # <class 'NoneType'>
````

---

## 🔹 3. Type Conversion

You can convert between types using built-in functions:
- `int()` → convert to integer
- `float()` → convert to float
- `str()` → convert to string
- `bool()` → convert to boolean

````python
# ▶️ Type conversion examples
x = "42"
print(type(x))         # str

x = int(x)
print(type(x))         # int
print(x + 8)           # 50

y = float("3.14")
print(y)               # 3.14

z = str(100)
print("Value: " + z)   # Value: 100
````

---

## 🔹 4. String Basics

Strings are one of the most used types in Python. Here are some key operations:

````python
# ▶️ String operations
greeting = "Hello, World!"

print(len(greeting))           # Length: 13
print(greeting.upper())        # HELLO, WORLD!
print(greeting.lower())        # hello, world!
print(greeting.replace("World", "Python"))  # Hello, Python!

# f-strings (modern way to format strings)
name = "Alice"
age = 25
print(f"My name is {name} and I am {age} years old.")
````

### Exercise 2: Type Conversion
Given the string `"100"`, convert it to:
1. An **integer** and add `50` to it
2. A **float**
3. A **boolean**

Print the result and type of each.

````python
# 🖊️ Your code here:
value = "100"

# 1. Convert to int and add 50
value_int = int(value) + 50

# 2. Convert to float
value_float = float(value)

# 3. Convert to boolean
value_bool = bool(int(value))

print("Integer:", value_int, type(value_int))
print("Float:", value_float, type(value_float))
print("Boolean:", value_bool, type(value_bool))
# ���️ Remember to convert the string back to an integer when performing arithmetic operations!
````
