# 📘 Lesson 2: Operators

Operators are symbols that perform operations on variables and values.
Python has several types of operators:

| Type | Symbols |
|------|---------|
| Arithmetic | `+`, `-`, `*`, `/`, `//`, `%`, `**` |
| Comparison | `==`, `!=`, `>`, `<`, `>=`, `<=` |
| Logical | `and`, `or`, `not` |
| Assignment | `=`, `+=`, `-=`, `*=`, `/=` |

---

## 🔹 1. Arithmetic Operators

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `+` | Addition | `5 + 3` | `8` |
| `-` | Subtraction | `5 - 3` | `2` |
| `*` | Multiplication | `5 * 3` | `15` |
| `/` | Division | `5 / 2` | `2.5` |
| `//` | Floor Division | `5 // 2` | `2` |
| `%` | Modulus (remainder) | `5 % 2` | `1` |
| `**` | Exponentiation | `5 ** 2` | `25` |

```python
# ▶️ Arithmetic operators in action
a = 10
b = 3

print(f"{a} + {b} = {a + b}")    # 13
print(f"{a} - {b} = {a - b}")    # 7
print(f"{a} * {b} = {a * b}")    # 30
print(f"{a} / {b} = {a / b}")    # 3.333...
print(f"{a} // {b} = {a // b}")  # 3  (floor division)
print(f"{a} % {b} = {a % b}")    # 1  (remainder)
print(f"{a} ** {b} = {a ** b}")  # 1000 (10 to the power of 3)
```

---

## 🔹 2. Comparison Operators

Comparison operators **compare two values** and always return a `bool` (`True` or `False`).

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `>` | Greater than | `5 > 3` | `True` |
| `<` | Less than | `5 < 3` | `False` |
| `>=` | Greater than or equal | `5 >= 5` | `True` |
| `<=` | Less than or equal | `5 <= 3` | `False` |

```python
# ▶️ Comparison operators in action
x = 10
y = 20

print(f"{x} == {y} →", x == y)   # False
print(f"{x} != {y} →", x != y)   # True
print(f"{x} > {y}  →", x > y)    # False
print(f"{x} < {y}  →", x < y)    # True
print(f"{x} >= {y} →", x >= y)   # False
print(f"{x} <= {y} →", x <= y)   # True
```

---

## 🔹 3. Logical Operators

Logical operators combine multiple conditions.

| Operator | Description | Example |
|----------|-------------|---------|
| `and` | True if **both** conditions are True | `x > 5 and x < 20` |
| `or` | True if **at least one** condition is True | `x < 5 or x > 8` |
| `not` | **Reverses** the result | `not(x > 5)` |

```python
# ▶️ Logical operators in action
age = 25
has_id = True

# and — both must be True
print(age >= 18 and has_id)       # True

# or — at least one must be True
print(age < 18 or has_id)         # True

# not — flips the result
print(not has_id)                  # False
print(not (age < 18))             # True
```

---

## 🔹 4. Assignment Operators

Shorthand ways to update a variable's value.

| Operator | Equivalent to |
|----------|---------------|
| `x += 5` | `x = x + 5` |
| `x -= 5` | `x = x - 5` |
| `x *= 5` | `x = x * 5` |
| `x /= 5` | `x = x / 5` |
| `x //= 5` | `x = x // 5` |
| `x %= 5` | `x = x % 5` |
| `x **= 5` | `x = x ** 5` |

```python
# ▶️ Assignment operators in action
score = 100

score += 10    # score = 110
print("After += 10:", score)

score -= 20    # score = 90
print("After -= 20:", score)

score *= 2     # score = 180
print("After *= 2:", score)

score //= 4    # score = 45
print("After //= 4:", score)

score **= 2    # score = 2025
print("After **= 2:", score)
```
