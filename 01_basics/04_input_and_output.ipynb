# 📘 Lesson 24: Input & Output

Input and output let your program **communicate with users**.

- **Input** means getting information from the user.
- **Output** means showing information to the user.

| Concept | Description |
|---------|-------------|
| `print()` | Displays output |
| `input()` | Gets user input as text |
| f-strings | Format output with variables |
| `int()` / `float()` | Convert input text to numbers |
| `sep` / `end` | Customize how `print()` behaves |

---

## 🔹 1. Basic Output with `print()`

Use `print()` to display text, numbers, variables, or results.

````python
# ▶️ Basic print examples
print("Hello, Python!")
print(42)
print("The answer is", 42)

name = "Alice"
print("Welcome", name)
````

---

## 🔹 2. Formatting Output with f-Strings

f-strings let you insert variables directly into text.

```python
name = "Alice"
print(f"Hello, {name}!")
```

````python
# ▶️ f-string output
name = "Alice"
age = 30
city = "New York"

print(f"My name is {name}.")
print(f"I am {age} years old.")
print(f"I live in {city}.")
print(f"Next year, I will be {age + 1} years old.")
````

---

## 🔹 3. Rounding Numbers in Output

You can format decimal numbers using f-strings.

```python
f"{number:.2f}"
```

This shows the number with 2 decimal places.

````python
# ▶️ Format decimal numbers
price = 19.9876
tax = 1.734
total = price + tax

print(f"Price: ${price:.2f}")
print(f"Tax: ${tax:.2f}")
print(f"Total: ${total:.2f}")
````

---

## 🔹 4. Getting Input with `input()`

Use `input()` to ask the user for information.

> 💡 In Jupyter notebooks, running an `input()` cell will show a text box where you can type your answer.

````python
# ▶️ Basic input
name = input("What is your name? ")
print(f"Hello, {name}!")
````

---

## 🔹 5. `input()` Always Returns a String

Even if the user types a number, `input()` gives you a string.

Use `int()` for whole numbers and `float()` for decimal numbers.

````python
# ▶️ Convert input to a number
age_text = input("Enter your age: ")
age = int(age_text)

print(f"You are {age} years old.")
print(f"Next year, you will be {age + 1}.")
````

---

## 🔹 6. Working with Decimal Input

Use `float()` when the input might contain decimals.

````python
# ▶️ Decimal input
price = float(input("Enter item price: "))
quantity = int(input("Enter quantity: "))

total = price * quantity
print(f"Total cost: ${total:.2f}")
````

---

## 🔹 7. Customizing `print()` with `sep`

`sep` controls what appears between multiple values in `print()`.

By default, `sep` is a space.

````python
# ▶️ Using sep
print("Python", "JavaScript", "SQL")
print("Python", "JavaScript", "SQL", sep=" | ")
print("2026", "05", "22", sep="-")
````

---

## 🔹 8. Customizing `print()` with `end`

`end` controls what happens after `print()` finishes.

By default, `end` is a newline character.

````python
# ▶️ Using end
print("Hello", end=" ")
print("Python!")

print("Loading", end="...")
print("Done")
````

## 🔹 9. Mini Project: Simple Calculator

This example combines input, conversion, calculations, and output.

````python
# ▶️ Simple calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print(f"{num1} + {num2} = {num1 + num2}")
print(f"{num1} - {num2} = {num1 - num2}")
print(f"{num1} * {num2} = {num1 * num2}")
print(f"{num1} / {num2} = {num1 / num2}")
````
