# 📘 Lesson 23: String Deep Dive

Strings are used to store and work with text. In this lesson, we will go deeper into indexing, slicing, methods, formatting, and common string patterns.

| Concept | Description |
|---------|-------------|
| Indexing | Get one character by position |
| Slicing | Get part of a string |
| Methods | Built-in string actions like `.upper()` and `.strip()` |
| f-strings | Format text with variables |
| Immutability | Strings cannot be changed directly |

---

## 🔹 1. Strings Are Sequences

A string is a sequence of characters. Each character has an index, starting at `0`. Negative indexes count from the end.

````python
# ▶️ String indexing
word = "Python"

print(word[0])    # P
print(word[1])    # y
print(word[-1])   # n
print(word[-2])   # o
````

---

## 🔹 2. Slicing Strings

Slicing lets you get part of a string.

```python
text[start:stop:step]
```

The `stop` index is not included.

````python
# ▶️ String slicing
message = "Learning Python"

print(message[0:8])    # Learning
print(message[:8])     # Learning
print(message[9:])     # Python
print(message[::2])    # every other character
print(message[::-1])   # reversed string
````

---

## 🔹 3. Strings Are Immutable

Strings cannot be changed directly after they are created. Instead, create a new string.

````python
# ▶️ String immutability
word = "Jython"

# word[0] = "P"  # ❌ This would cause an error

fixed_word = "P" + word[1:]
print(fixed_word)
````

---

## 🔹 4. Common String Methods

String methods return new strings or useful information.

| Method | Description |
|--------|-------------|
| `.upper()` | Convert to uppercase |
| `.lower()` | Convert to lowercase |
| `.title()` | Capitalize each word |
| `.strip()` | Remove spaces from both ends |
| `.replace()` | Replace text |
| `.count()` | Count occurrences |

````python
# ▶️ Common string methods
text = "  python is fun  "

print(text.upper())
print(text.lower())
print(text.title())
print(text.strip())
print(text.replace("fun", "powerful"))
print(text.count("n"))
````

---

## 🔹 5. Searching Inside Strings

Use `in` to check if text exists inside a string. Use `.find()` to get the index of the first match.

````python
# ▶️ Search strings
sentence = "Python makes programming enjoyable."

print("Python" in sentence)
print("Java" in sentence)
print(sentence.find("programming"))
print(sentence.find("missing"))  # -1 means not found
````

---

## 🔹 6. Splitting and Joining Strings

`.split()` turns a string into a list. `.join()` combines a list of strings into one string.

````python
# ▶️ Split and join
shopping_text = "apples,bananas,carrots"
items = shopping_text.split(",")
print(items)

clean_text = " | ".join(items)
print(clean_text)
````

---

## 🔹 7. f-Strings and Formatting

f-strings make it easy to insert variables and expressions into text.

````python
# ▶️ f-string examples
name = "Alice"
score = 92.567

print(f"Student: {name}")
print(f"Score: {score}")
print(f"Rounded score: {score:.2f}")
print(f"Passed: {score >= 60}")
````

---

## 🔹 8. Useful Checks

Python has methods that check what kind of characters a string contains.

````python
# ▶️ String check methods
print("Python".isalpha())
print("12345".isdigit())
print("Python3".isalnum())
print("   ".isspace())
print("hello".islower())
print("HELLO".isupper())
````
