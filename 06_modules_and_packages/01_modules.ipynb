# 📘 Lesson 12: Modules & Packages

Modules help you organize Python code into reusable files. Packages help you organize multiple modules into folders.

| Concept | Description |
|---------|-------------|
| Module | A Python file you can import, like `math` |
| Package | A folder containing modules |
| Import | Loading code from another module |
| Standard library | Built-in Python modules you can use immediately |

---

## 🔹 1. What Is a Module?

A **module** is a Python file that contains code such as functions, variables, or classes.

Python includes many ready-to-use modules in the **standard library**.

````python
# ▶️ Importing a standard library module
import math

print(math.sqrt(25))      # square root
print(math.pi)            # pi value
print(math.ceil(4.2))     # round up
````

---

## 🔹 2. Importing Specific Things

Instead of importing the whole module, you can import only what you need.

````python
# ▶️ Import specific functions
from math import sqrt, pow

print(sqrt(64))
print(pow(2, 3))
````

---

## 🔹 3. Import Aliases

An alias gives a module a shorter name. This is common with long module names.

````python
# ▶️ Import with an alias
import random as rnd

print(rnd.randint(1, 10))
print(rnd.choice(["apple", "banana", "cherry"]))
````

---

## 🔹 4. Useful Standard Library Modules

Python comes with many helpful modules already installed.

````python
# ▶️ datetime module
import datetime

today = datetime.date.today()
now = datetime.datetime.now()

print(f"Today: {today}")
print(f"Current time: {now.strftime('%H:%M:%S')}")
````

````python
# ▶️ os module
import os

print("Current folder:", os.getcwd())
print("Files here:", os.listdir(".")[:5])
````

---

## 🔹 5. Creating Your Own Module

To create your own module, make a `.py` file and put functions inside it.

Example file: `calculator.py`
```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b
```

Then you can import it:
```python
import calculator
print(calculator.add(5, 3))
```

> ⚠️ In notebooks, creating local `.py` files depends on your folder setup. The idea is the same in normal Python projects.

---

## 🔹 6. What Is a Package?

A **package** is a folder that groups related modules together.

Example structure:
```
my_package/
    __init__.py
    calculator.py
    greetings.py
```

You can then import from the package:
```python
from my_package import calculator
```

---

## 🔹 7. Installing Third-Party Packages

Some packages are not built into Python. You install them using `pip`.

```bash
pip install package-name
```

Example:
```bash
pip install requests
```

> 💡 We avoid external dependencies in most beginner lessons, but `pip` is how real projects add extra tools.
