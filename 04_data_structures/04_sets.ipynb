# 📘 Lesson 9: Sets

A **set** is an unordered collection of unique items.
Sets are useful when you want to remove duplicates, check membership quickly, or compare groups of data.

| Feature | Description |
|---------|-------------|
| ❌ Unordered | Items do not have a fixed position |
| ✅ Mutable | You can add and remove items |
| ✅ Unique items | Duplicates are automatically removed |
| ❌ No indexing | You cannot access items with `[0]` |

---

## 🔹 1. Creating Sets

````python
# ▶️ Creating sets
fruits = {"apple", "banana", "cherry"}
numbers = {1, 2, 3, 4, 5}
mixed = {"Alice", 30, True, 3.14}

# Duplicates are removed automatically
duplicates = {1, 2, 2, 3, 3, 3, 4}

# Empty set: use set(), not {}
empty_set = set()
empty_dict = {}

print(fruits)
print(numbers)
print(mixed)
print(duplicates)
print(type(empty_set))   # set
print(type(empty_dict))  # dict
````

---

## 🔹 2. Sets Are Unordered

Sets do not keep items in a reliable index order.
Because of this, you cannot access set items using indexes like `my_set[0]`.

````python
# ▶️ Sets have no indexing
colors = {"red", "green", "blue"}

print(colors)

# print(colors[0])  # ❌ TypeError: 'set' object is not subscriptable

# Use a loop to access items
for color in colors:
    print(color)
````

---

## 🔹 3. Adding Items

````python
# ▶️ Adding items to a set
fruits = {"apple", "banana"}

fruits.add("cherry")                 # add one item
print("After add:", fruits)

fruits.update(["mango", "orange"])  # add multiple items
print("After update:", fruits)

fruits.add("apple")                  # duplicate ignored
print("After duplicate add:", fruits)
````

---

## 🔹 4. Removing Items

| Method | Description |
|--------|-------------|
| `set.remove(x)` | Removes x, raises an error if missing |
| `set.discard(x)` | Removes x, does nothing if missing ✅ safer |
| `set.pop()` | Removes and returns a random item |
| `set.clear()` | Removes all items |

````python
# ▶️ Removing items from a set
fruits = {"apple", "banana", "cherry", "mango"}

fruits.remove("banana")
print("After remove:", fruits)

fruits.discard("pineapple")  # no error if missing
print("After discard missing item:", fruits)

popped = fruits.pop()
print("Popped:", popped)
print("After pop:", fruits)

fruits.clear()
print("After clear:", fruits)
````

---

## 🔹 5. Checking Membership

````python
# ▶️ Membership checks
allowed_users = {"alice", "bob", "charlie"}

print("alice" in allowed_users)     # True
print("diana" in allowed_users)     # False
print("admin" not in allowed_users) # True
````

---

## 🔹 6. Set Operations

Sets are great for comparing groups.

| Operation | Method | Symbol | Meaning |
|-----------|--------|--------|---------|
| Union | `a.union(b)` | `a | b` | Items in either set |
| Intersection | `a.intersection(b)` | `a & b` | Items in both sets |
| Difference | `a.difference(b)` | `a - b` | Items in a but not b |
| Symmetric difference | `a.symmetric_difference(b)` | `a ^ b` | Items in either, but not both |

````python
# ▶️ Set operations
python_students = {"Alice", "Bob", "Charlie", "Diana"}
javascript_students = {"Charlie", "Diana", "Eve", "Frank"}

print("Union:", python_students | javascript_students)
print("Intersection:", python_students & javascript_students)
print("Python only:", python_students - javascript_students)
print("One course only:", python_students ^ javascript_students)
````

---

## 🔹 7. Useful Set Methods

````python
# ▶️ Useful set methods
a = {1, 2, 3}
b = {1, 2, 3, 4, 5}
c = {10, 20}

print("Length of b:", len(b))
print("a is subset of b:", a.issubset(b))
print("b is superset of a:", b.issuperset(a))
print("a and c are disjoint:", a.isdisjoint(c))
````

---

## 🔹 8. Set Comprehension

Set comprehension is like list comprehension, but creates a set.

```python
new_set = {expression for item in iterable if condition}
```

````python
# ▶️ Set comprehension examples

# Squares of numbers 1 to 5
squares = {x ** 2 for x in range(1, 6)}
print("Squares:", squares)

# Unique first letters
words = ["apple", "banana", "apricot", "blueberry", "cherry"]
first_letters = {word[0] for word in words}
print("First letters:", first_letters)

# Even numbers only
evens = {x for x in range(1, 21) if x % 2 == 0}
print("Evens:", evens)
````
