# 📘 Lesson 8: Tuples

A **tuple** is an ordered, immutable collection of items.
Tuples are similar to lists, but once created, their items cannot be changed.

| Feature | Description |
|---------|-------------|
| ✅ Ordered | Items have a fixed position (index) |
| ❌ Immutable | You cannot add, remove, or change items directly |
| ✅ Mixed types | Can hold strings, ints, bools, lists, and more |
| ✅ Duplicates | Same value can appear multiple times |

---

## 🔹 1. Creating Tuples

```python
# ▶️ Creating tuples
fruits = ("apple", "banana", "cherry")
numbers = (1, 2, 3, 4, 5)
mixed = ("Alice", 30, True, 3.14)
empty = ()
nested = ((1, 2), (3, 4), (5, 6))

# A one-item tuple needs a comma!
single = ("only item",)
not_tuple = ("only item")

print(fruits)
print(numbers)
print(mixed)
print(f"Length: {len(fruits)}")
print(type(single))      # tuple
print(type(not_tuple))   # str
```

---

## 🔹 2. Indexing & Slicing

Tuples use the same indexing and slicing rules as lists.
Indexing starts at `0`, and negative indexes count from the end.

```python
# ▶️ Indexing and slicing
colors = ("red", "green", "blue", "yellow", "purple")

print(colors[0])     # red
print(colors[2])     # blue
print(colors[-1])    # purple
print(colors[1:4])   # ('green', 'blue', 'yellow')
print(colors[:3])    # ('red', 'green', 'blue')
print(colors[::-1])  # reversed tuple
```

---

## 🔹 3. Tuples Are Immutable

Immutable means you **cannot change the tuple itself** after it is created.
This makes tuples useful for data that should stay fixed, like coordinates, dates, or settings.

```python
# ▶️ Immutability example
point = (10, 20)
print(point)

# point[0] = 99  # ❌ TypeError: 'tuple' object does not support item assignment

# To 'change' a tuple, create a new tuple
point = (99, point[1])
print(point)
```

---

## 🔹 4. Tuple Methods

Because tuples cannot be changed, they have fewer methods than lists.

| Method / Function | Description |
|-------------------|-------------|
| `len(tuple)` | Number of items |
| `tuple.count(x)` | Count occurrences of x |
| `tuple.index(x)` | Index of first occurrence of x |
| `x in tuple` | Check if x exists (True/False) |
| `min(tuple)` | Smallest value |
| `max(tuple)` | Largest value |
| `sum(tuple)` | Sum of numeric values |

```python
# ▶️ Tuple methods and built-in functions
numbers = (4, 2, 9, 1, 7, 2, 5, 2)

print("Length:", len(numbers))
print("Count of 2:", numbers.count(2))
print("Index of 9:", numbers.index(9))
print("7 in tuple?", 7 in numbers)
print("10 in tuple?", 10 in numbers)
print("Min:", min(numbers))
print("Max:", max(numbers))
print("Sum:", sum(numbers))
```

---

## 🔹 5. Tuple Packing & Unpacking

Packing means putting multiple values into one tuple.
Unpacking means assigning tuple values back into separate variables.

```python
# ▶️ Packing and unpacking
person = ("Alice", 30, "London")   # packing

name, age, city = person                 # unpacking

print(name)
print(age)
print(city)

# Extended unpacking
numbers = (1, 2, 3, 4, 5)
first, second, *rest = numbers
print("First:", first)
print("Second:", second)
print("Rest:", rest)      # rest becomes a list
```

---

## 🔹 6. Looping Through Tuples

```python
# ▶️ Loop through a tuple
languages = ("Python", "JavaScript", "Go", "Rust")

for language in languages:
    print(language)

print()

for i, language in enumerate(languages):
    print(f"{i + 1}. {language}")
```

---

## 🔹 7. Tuple ↔ List Conversion

If you need to modify tuple data, you can convert it to a list, change it, then convert it back.

```python
# ▶️ Convert tuple to list and back
fruits = ("apple", "banana", "cherry")

fruit_list = list(fruits)
fruit_list.append("mango")
fruit_list[1] = "blueberry"

fruits = tuple(fruit_list)
print(fruits)
```

---

## 🔹 8. When Should You Use Tuples?

Use a tuple when:
- The data should **not change**
- The order matters
- You want to group related values together
- You are returning multiple values from a function

Common examples: coordinates, RGB colors, dates, database rows, and fixed configuration values.

```python
# ▶️ Practical tuple examples
coordinate = (12.34, 56.78)
rgb_color = (255, 128, 0)
date = (2026, 5, 22)

def get_name_and_score():
    return "Alice", 95   # returns a tuple automatically

student, score = get_name_and_score()

print("Coordinate:", coordinate)
print("RGB color:", rgb_color)
print("Date:", date)
print(f"{student} scored {score}")
```

```python
# ✅ Solution - Exercise 3: Tuple Stats
def tuple_stats(numbers):
    return {
        "count": len(numbers),
        "sum": sum(numbers),
        "average": round(sum(numbers) / len(numbers), 2),
        "min": min(numbers),
        "max": max(numbers)
    }

print(tuple_stats((15, 42, 8, 23, 16, 4, 42, 9)))
```

```python
# ✅ Solution - Exercise 4: Update a Tuple Safely
def replace_item(tpl, old, new):
    items = list(tpl)
    if old in items:
        index = items.index(old)
        items[index] = new
    return tuple(items)

print(replace_item(("red", "green", "blue"), "green", "yellow"))
# ('red', 'yellow', 'blue')
```
