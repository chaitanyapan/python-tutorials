# 📘 Lesson 6: Lists

A **list** is an ordered, mutable collection of items.
It's one of the most used data structures in Python.

| Feature | Description |
|---------|-------------|
| ✅ Ordered | Items have a fixed position (index) |
| ✅ Mutable | You can add, remove, change items |
| ✅ Mixed types | Can hold strings, ints, bools, even other lists |
| ✅ Duplicates | Same value can appear multiple times |

---

## 🔹 1. Creating Lists

````python
# ▶️ Creating lists
fruits = ["apple", "banana", "cherry"]
numbers = [1, 2, 3, 4, 5]
mixed = ["Alice", 30, True, 3.14]    # mixed types
empty = []                            # empty list
nested = [[1, 2], [3, 4], [5, 6]]    # list of lists

print(fruits)
print(numbers)
print(mixed)
print(f"Length: {len(fruits)}")
````

## 🔹 2. Indexing & Slicing

Access items using their **index** (starts at `0`).
Negative indexes count from the **end**.

```
 List:   ["apple", "banana", "cherry", "date", "elderberry"]
 Index:      0        1         2        3           4
 Neg:        -5       -4        -3       -2          -1
```

````python
# ▶️ Indexing
fruits = ["apple", "banana", "cherry", "date", "elderberry"]

print(fruits[0])    # apple   (first)
print(fruits[2])    # cherry
print(fruits[-1])   # elderberry (last)
print(fruits[-2])   # date
````

````python
# ▶️ Slicing — list[start:stop:step]  (stop is exclusive)
fruits = ["apple", "banana", "cherry", "date", "elderberry"]

print(fruits[1:3])    # ['banana', 'cherry']
print(fruits[:3])     # ['apple', 'banana', 'cherry']  (from start)
print(fruits[2:])     # ['cherry', 'date', 'elderberry'] (to end)
print(fruits[::2])    # every other item
print(fruits[::-1])   # reversed!
````

---

## 🔹 3. Modifying Lists

Lists are **mutable** — you can change them after creation.

````python
# ▶️ Changing items
fruits = ["apple", "banana", "cherry"]
fruits[1] = "mango"    # replace banana with mango
print(fruits)           # ['apple', 'mango', 'cherry']
````

````python
# ▶️ Adding items
fruits = ["apple", "banana"]

fruits.append("cherry")          # add to end
print("append:", fruits)

fruits.insert(1, "mango")        # insert at index 1
print("insert:", fruits)

fruits.extend(["date", "elderberry"])  # add multiple items
print("extend:", fruits)
````

````python
# ▶️ Removing items
fruits = ["apple", "banana", "cherry", "banana", "date"]

fruits.remove("banana")   # removes FIRST occurrence
print("remove:", fruits)

popped = fruits.pop()     # removes & returns last item
print("popped:", popped)
print("after pop:", fruits)

popped_index = fruits.pop(1)   # removes at index 1
print("popped at 1:", popped_index)
print("after pop(1):", fruits)

fruits.clear()            # removes everything
print("cleared:", fruits)
````

## 🔹 4. Useful List Methods

| Method | Description |
|--------|-------------|
| `len(list)` | Number of items |
| `list.sort()` | Sort in place (ascending) |
| `sorted(list)` | Return new sorted list |
| `list.reverse()` | Reverse in place |
| `list.count(x)` | Count occurrences of x |
| `list.index(x)` | Index of first occurrence of x |
| `x in list` | Check if x exists (True/False) |

````python
# ▶️ List methods
numbers = [4, 2, 9, 1, 7, 2, 5, 2]

print("Length:", len(numbers))
print("Count of 2:", numbers.count(2))
print("Index of 9:", numbers.index(9))
print("7 in list?", 7 in numbers)
print("10 in list?", 10 in numbers)

numbers.sort()
print("Sorted:", numbers)

numbers.reverse()
print("Reversed:", numbers)
````

## 🔹 5. Looping Through Lists

````python
# ▶️ Loop through a list
fruits = ["apple", "banana", "cherry"]

# Simple loop
for fruit in fruits:
    print(fruit)

print()

# With index using enumerate
for i, fruit in enumerate(fruits):
    print(f"{i + 1}. {fruit}")
````

## 🔹 6. List Comprehension

A powerful one-liner to **create lists** from existing ones.

```python
new_list = [expression for item in iterable if condition]
```

````python
# ▶️ List comprehension examples

# Squares of 1 to 10
squares = [x ** 2 for x in range(1, 11)]
print("Squares:", squares)

# Even numbers only
evens = [x for x in range(1, 21) if x % 2 == 0]
print("Evens:", evens)

# Uppercase all fruits
fruits = ["apple", "banana", "cherry"]
upper_fruits = [f.upper() for f in fruits]
print("Uppercase:", upper_fruits)

# Filter fruits with more than 5 letters
long_fruits = [f for f in fruits if len(f) > 5]
print("Long fruits:", long_fruits)
````

````python
# ✅ Solution - Exercise 4: Remove Duplicates
def remove_duplicates(lst):
    result = []
    for item in lst:
        if item not in result:
            result.append(item)
    return result

print(remove_duplicates([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]))
# [3, 1, 4, 5, 9, 2, 6]
````
