# 📘 Lesson 7: Dictionaries

A **dictionary** stores data as **key-value pairs** — like a real dictionary where you look up a word (key) to get its definition (value).

| Feature | Description |
|---------|-------------|
| ✅ Key-Value pairs | Each item has a unique key and a value |
| ✅ Ordered | Insertion order is preserved (Python 3.7+) |
| ✅ Mutable | Add, update, remove items |
| ❌ Duplicate keys | Keys must be unique (values can repeat) |

```python
person = {
    "name": "Alice",
    "age": 30,
    "city": "London"
}
#         ↑ key    ↑ value
```

---

## 🔹 1. Creating Dictionaries

````python
# ▶️ Creating dictionaries
person = {
    "name": "Alice",
    "age": 30,
    "city": "London"
}

# Using dict() constructor
car = dict(brand="Toyota", model="Corolla", year=2022)

empty = {}   # empty dictionary

print(person)
print(car)
print(f"Keys: {len(person)}")
````

---

## 🔹 2. Accessing Values

Two ways to access values:
- `dict[key]` → raises `KeyError` if key doesn't exist
- `dict.get(key)` → returns `None` (or a default) if key doesn't exist ✅ safer!

````python
# ▶️ Accessing values
person = {"name": "Alice", "age": 30, "city": "London"}

# Using [] — raises KeyError if missing
print(person["name"])    # Alice
print(person["age"])     # 30

# Using .get() — safe, returns None if missing
print(person.get("city"))        # London
print(person.get("email"))       # None
print(person.get("email", "N/A")) # N/A  ← custom default!
````

---

## 🔹 3. Adding & Updating Items

````python
# ▶️ Adding & updating
person = {"name": "Alice", "age": 30}

# Add new key
person["email"] = "alice@example.com"
print("After add:", person)

# Update existing key
person["age"] = 31
print("After update:", person)

# Update multiple keys at once
person.update({"city": "Paris", "age": 32})
print("After update():", person)
````

---

## 🔹 4. Removing Items

| Method | Description |
|--------|-------------|
| `del dict[key]` | Remove by key |
| `dict.pop(key)` | Remove & return value |
| `dict.popitem()` | Remove & return last inserted pair |
| `dict.clear()` | Remove all items |

````python
# ▶️ Removing items
person = {"name": "Alice", "age": 30, "city": "London", "email": "alice@example.com"}

del person["email"]
print("After del:", person)

age = person.pop("age")    # removes and returns the value
print(f"Popped age: {age}")
print("After pop:", person)

last = person.popitem()    # removes last inserted key-value pair
print(f"Popped last: {last}")
print("After popitem:", person)
````

---

## 🔹 5. Looping Through Dictionaries

````python
# ▶️ Looping through a dictionary
person = {"name": "Alice", "age": 30, "city": "London"}

# Loop through keys (default)
print("--- Keys ---")
for key in person:
    print(key)

# Loop through values
print("--- Values ---")
for value in person.values():
    print(value)

# Loop through key-value pairs (most common!)
print("--- Key-Value pairs ---")
for key, value in person.items():
    print(f"{key}: {value}")
````

---

## 🔹 6. Checking Keys & Useful Methods

````python
# ▶️ Useful dictionary methods
person = {"name": "Alice", "age": 30, "city": "London"}

# Check if key exists
print("name" in person)     # True
print("email" in person)    # False

# Get all keys, values, items
print(list(person.keys()))    # ['name', 'age', 'city']
print(list(person.values()))  # ['Alice', 30, 'London']
print(list(person.items()))   # [('name', 'Alice'), ...]

# Copy a dictionary
person_copy = person.copy()
person_copy["name"] = "Bob"   # doesn't affect original
print("Original:", person)
print("Copy:", person_copy)
````

---

## 🔹 7. Nested Dictionaries

Dictionaries can contain other dictionaries — great for representing structured data.

````python
# ▶️ Nested dictionaries
students = {
    "alice": {"age": 20, "grade": "A", "score": 95},
    "bob":   {"age": 22, "grade": "B", "score": 82},
    "charlie": {"age": 21, "grade": "A", "score": 91},
}

# Access nested value
print(students["alice"]["score"])     # 95
print(students["bob"]["grade"])       # B

# Loop through nested dict
for name, info in students.items():
    print(f"{name.capitalize()}: Grade={info['grade']}, Score={info['score']}")
````

---

## 🔹 8. Dictionary Comprehension

Just like list comprehension, but for dictionaries!

```python
new_dict = {key: value for item in iterable if condition}
```

````python
# ▶️ Dictionary comprehension

# Squares dictionary: {1: 1, 2: 4, 3: 9, ...}
squares = {x: x**2 for x in range(1, 6)}
print("Squares:", squares)

# Filter — only students with score >= 90
students = {"alice": 95, "bob": 82, "charlie": 91, "diana": 78}
top_students = {name: score for name, score in students.items() if score >= 90}
print("Top students:", top_students)

# Swap keys and values
original = {"a": 1, "b": 2, "c": 3}
swapped = {v: k for k, v in original.items()}
print("Swapped:", swapped)
````

````python
# ✅ Solution - Exercise 1: Contact Book
contacts = {
    "Alice": {"phone": "555-1234", "email": "alice@email.com"},
    "Bob":   {"phone": "555-5678", "email": "bob@email.com"},
    "Charlie": {"phone": "555-9012", "email": "charlie@email.com"},
}

# 1. All names
print("Contacts:", list(contacts.keys()))

# 2. Alice's phone
print("Alice's phone:", contacts["Alice"]["phone"])

# 3. Add new contact
contacts["Diana"] = {"phone": "555-3456", "email": "diana@email.com"}

# 4. Update email
contacts["Bob"]["email"] = "bob.new@email.com"

# 5. Print all
print("\n--- All Contacts ---")
for name, info in contacts.items():
    print(f"{name}: 📞 {info['phone']} | ✉️  {info['email']}")
````

````python
# ✅ Solution - Exercise 2: Word Counter
def word_count(sentence):
    counts = {}
    for word in sentence.split():
        counts[word] = counts.get(word, 0) + 1
    return counts

print(word_count("the cat sat on the mat the cat"))
````

````python
# ✅ Solution - Exercise 3: Gradebook
def summarize_grades(gradebook):
    best_student = None
    best_avg = 0

    for student, scores in gradebook.items():
        avg = round(sum(scores) / len(scores), 1)

        if avg >= 90:   grade = "A"
        elif avg >= 80: grade = "B"
        elif avg >= 70: grade = "C"
        elif avg >= 60: grade = "D"
        else:           grade = "F"

        passed = "✅ Passed" if avg >= 60 else "❌ Failed"
        print(f"{student}: Avg={avg}, Grade={grade}, {passed}")

        if avg > best_avg:
            best_avg = avg
            best_student = student

    print(f"\n🏆 Highest average: {best_student} ({best_avg})")

gradebook = {
    "Alice":   [85, 92, 78, 95, 88],
    "Bob":     [70, 65, 80, 75, 72],
    "Charlie": [90, 95, 92, 98, 94],
    "Diana":   [60, 55, 70, 65, 58]
}
summarize_grades(gradebook)
````

````python
# ✅ Solution - Exercise 4: Inventory Manager
def update_inventory(inventory, sold):
    updated = inventory.copy()
    for item, qty in sold.items():
        if item in updated:
            updated[item] -= qty
    # Remove items with 0 or less stock
    updated = {item: qty for item, qty in updated.items() if qty > 0}
    return updated

inventory = {"apples": 50, "bananas": 30, "oranges": 20, "grapes": 10}
sold = {"apples": 20, "bananas": 30, "grapes": 15}
print(update_inventory(inventory, sold))
# {'apples': 30, 'oranges': 20}
````
