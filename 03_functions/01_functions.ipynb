# 📘 Lesson 5: Functions

A **function** is a reusable block of code that performs a specific task.
Instead of writing the same code repeatedly, you define it once and **call it whenever needed**.

### Why use functions?
- ♻️ **Reusability** — write once, use many times
- 🧹 **Cleaner code** — break big problems into small pieces
- 🐛 **Easier debugging** — fix in one place, fixed everywhere

---

## 🔹 1. Defining & Calling a Function

```python
def function_name():
    # code block
```

- `def` keyword defines the function
- The function body must be **indented**
- Call it by writing `function_name()`

````python
# ▶️ Define and call a simple function
def greet():
    print("Hello! Welcome to Python.")

# Calling the function
greet()
greet()   # call it as many times as you want!
````

---

## 🔹 2. Parameters & Arguments

Functions can accept **inputs** called parameters.

```python
def function_name(parameter1, parameter2):
    # use parameters inside
```

- **Parameter** → variable in the function definition
- **Argument** → actual value you pass when calling

````python
# ▶️ Function with parameters
def greet(name):
    print(f"Hello, {name}! Welcome to Python.")

greet("Alice")    # argument = "Alice"
greet("Bob")
greet("Charlie")
````

````python
# ▶️ Multiple parameters
def describe_person(name, age, city):
    print(f"{name} is {age} years old and lives in {city}.")

describe_person("Alice", 30, "New York")
describe_person("Bob", 25, "London")
````

---

## 🔹 3. Default Parameter Values

You can set a **default value** for a parameter — used when no argument is passed.

````python
# ▶️ Default parameter values
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet("Alice")                  # uses default greeting
greet("Bob", "Good morning")    # overrides default
greet("Charlie", "Hey")         # overrides default
````

---

## 🔹 4. `return` Statement

Functions can **send back a result** using `return`.
Without `return`, a function returns `None` by default.

> 💡 `print()` displays output. `return` sends a value back to the caller — they are very different!

````python
# ▶️ Function with return
def add(a, b):
    return a + b

result = add(5, 3)
print("5 + 3 =", result)    # 8

# You can use the returned value directly
print("10 + 20 =", add(10, 20))
````

````python
# ▶️ Returning multiple values
def min_max(numbers):
    return min(numbers), max(numbers)

nums = [4, 2, 9, 1, 7, 5]
smallest, largest = min_max(nums)

print(f"Min: {smallest}, Max: {largest}")
````

---

## 🔹 5. Keyword Arguments

You can pass arguments **by name** — order doesn't matter then!

````python
# ▶️ Keyword arguments
def create_profile(name, age, city):
    print(f"Name: {name} | Age: {age} | City: {city}")

# Positional (order matters)
create_profile("Alice", 30, "Paris")

# Keyword (order doesn't matter)
create_profile(city="Tokyo", name="Bob", age=25)
````

---

## 🔹 6. Variable Scope

- **Local variable** — defined inside a function, only accessible there
- **Global variable** — defined outside, accessible everywhere

````python
# ▶️ Scope example
language = "Python"   # global variable

def show_language():
    version = 3.12     # local variable — only exists inside this function
    print(f"Language: {language}, Version: {version}")

show_language()

print(language)   # works fine — global
# print(version)  # ❌ would raise NameError — version is local!
````

---

## 🔹 7. Lambda Functions (Anonymous Functions)

A **lambda** is a short, one-line function with no name.

```python
lambda parameters: expression
```

> 💡 Use lambdas for simple, throwaway functions. For anything complex, use `def`.

````python
# ▶️ Lambda examples

# Regular function
def square(x):
    return x ** 2

# Same thing as lambda
square_lambda = lambda x: x ** 2

print(square(5))         # 25
print(square_lambda(5))  # 25

# Lambda with multiple parameters
add = lambda a, b: a + b
print(add(3, 7))         # 10

# Practical use — sorting a list of tuples by second element
students = [("Alice", 90), ("Bob", 75), ("Charlie", 85)]
students.sort(key=lambda s: s[1])
print(students)
````

# ✅ Solutions

> 💡 **Try the exercises yourself first before looking at the solutions!**

````python
# ✅ Solution - Exercise 1: Temperature Converter
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_to_celsius(f):
    return (f - 32) * 5/9

print(celsius_to_fahrenheit(0))      # 32.0
print(fahrenheit_to_celsius(98.6))   # 37.0
````

````python
# ✅ Solution - Exercise 2: Calculator
def calculate(a, b, operation):
    if operation == "add":
        return a + b
    elif operation == "subtract":
        return a - b
    elif operation == "multiply":
        return a * b
    elif operation == "divide":
        if b == 0:
            return "Cannot divide by zero"
        return a / b
    else:
        return "Unknown operation"

print(calculate(10, 5, "add"))       # 15
print(calculate(10, 5, "subtract"))  # 5
print(calculate(10, 5, "multiply"))  # 50
print(calculate(10, 5, "divide"))    # 2.0
print(calculate(10, 0, "divide"))    # Cannot divide by zero
print(calculate(10, 5, "power"))     # Unknown operation
````

````python
# ✅ Solution - Exercise 3: Is Prime?
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

primes = [n for n in range(1, 51) if is_prime(n)]
print("Primes between 1 and 50:", primes)
````

````python
# ✅ Solution - Exercise 4: Grade Summary
def grade_summary(scores):
    avg = round(sum(scores) / len(scores), 2)

    if avg >= 90:
        grade = "A"
    elif avg >= 80:
        grade = "B"
    elif avg >= 70:
        grade = "C"
    elif avg >= 60:
        grade = "D"
    else:
        grade = "F"

    return {
        "average": avg,
        "highest": max(scores),
        "lowest": min(scores),
        "grade": grade
    }

result = grade_summary([88, 92, 75, 60, 95, 70])
print(result)
````
