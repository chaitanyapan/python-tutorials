# 📘 Lesson 27: `*args` and `**kwargs`

Sometimes you do not know ahead of time how many arguments a function should accept. Python gives us `*args` and `**kwargs` for flexible functions.

| Feature | Meaning | Example Use |
|---------|---------|-------------|
| `*args` | Accept many positional arguments | `add_all(1, 2, 3)` |
| `**kwargs` | Accept many keyword arguments | `profile(name="Alice", age=30)` |
| Positional argument | Passed by position | `greet("Alice")` |
| Keyword argument | Passed by name | `greet(name="Alice")` |

---

## 🔹 1. Regular Function Arguments

Before using flexible arguments, remember how normal function parameters work.

````python
# ▶️ Regular function arguments
def introduce(name, age):
    print(f"My name is {name} and I am {age} years old.")

introduce("Alice", 30)
introduce(name="Bob", age=25)
````

---

## 🔹 2. What Is `*args`?

`*args` lets a function accept any number of positional arguments. Inside the function, `args` behaves like a tuple.

> 💡 The name `args` is a convention. The `*` is what makes it work.

````python
# ▶️ Basic *args example
def show_items(*args):
    print(args)
    print(type(args))

show_items("apple", "banana", "cherry")
````

---

## 🔹 3. Looping Through `*args`

Because `args` is like a tuple, you can loop through it.

````python
# ▶️ Loop through *args
def print_names(*args):
    for name in args:
        print(f"Hello, {name}!")

print_names("Alice", "Bob", "Charlie")
````

---

## 🔹 4. Using `*args` for Calculations

`*args` is useful when you want to calculate with any number of values.

````python
# ▶️ Add any number of values
def add_all(*numbers):
    total = 0

    for number in numbers:
        total += number

    return total

print(add_all(1, 2, 3))
print(add_all(10, 20, 30, 40))
````

---

## 🔹 5. What Is `**kwargs`?

`**kwargs` lets a function accept any number of keyword arguments. Inside the function, `kwargs` behaves like a dictionary.

> 💡 The name `kwargs` means keyword arguments. The `**` is what makes it work.

````python
# ▶️ Basic **kwargs example
def show_profile(**kwargs):
    print(kwargs)
    print(type(kwargs))

show_profile(name="Alice", age=30, city="New York")
````

---

## 🔹 6. Looping Through `**kwargs`

Because `kwargs` is like a dictionary, you can loop through its keys and values.

````python
# ▶️ Loop through **kwargs
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_details(name="Bob", job="Developer", language="Python")
````

---

## 🔹 7. Combining Regular Arguments with `*args`

Regular parameters can come before `*args`. This is useful when one argument is required and the rest are flexible.

````python
# ▶️ Regular argument plus *args
def order_food(main_item, *extras):
    print(f"Main item: {main_item}")

    if extras:
        print("Extras:")
        for extra in extras:
            print(f"- {extra}")
    else:
        print("No extras.")

order_food("Burger", "cheese", "fries", "sauce")
````

---

## 🔹 8. Combining `*args` and `**kwargs`

You can use both in the same function. Put `*args` before `**kwargs`.

```python
def function_name(*args, **kwargs):
    ...
```

````python
# ▶️ Use *args and **kwargs together
def build_report(*scores, **student_info):
    print("Student Info:")
    for key, value in student_info.items():
        print(f"{key}: {value}")

    average = sum(scores) / len(scores)
    print(f"Average score: {average:.2f}")

build_report(90, 85, 92, name="Alice", course="Python")
````

---

## 🔹 9. Unpacking with `*` and `**`

You can also use `*` and `**` when calling a function to unpack a list, tuple, or dictionary.

````python
# ▶️ Unpacking arguments
def introduce(name, age, city):
    print(f"{name} is {age} years old and lives in {city}.")

person_list = ["Charlie", 28, "London"]
person_dict = {"name": "Diana", "age": 35, "city": "Paris"}

introduce(*person_list)
introduce(**person_dict)
````
