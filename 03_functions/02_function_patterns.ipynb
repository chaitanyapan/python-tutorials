# 📘 Lesson 26: Function Patterns

Function patterns are common ways to organize reusable code. Once you recognize these patterns, writing functions becomes much easier.

| Pattern | Description |
|---------|-------------|
| Input → Output | Take values in, return a result |
| Validation | Check if data is acceptable |
| Transformation | Convert data into a new form |
| Aggregation | Combine many values into one result |
| Helper function | Small function used by another function |

---

## 🔹 1. Input → Output Functions

A common function pattern is: take input, calculate something, and return output.

````python
# ▶️ Input → output pattern
def calculate_area(width, height):
    return width * height

area = calculate_area(5, 3)
print(f"Area: {area}")
````

---

## 🔹 2. Boolean Check Functions

Some functions answer a yes/no question by returning `True` or `False`. These often start with words like `is_`, `has_`, or `can_`.

````python
# ▶️ Boolean check pattern
def is_even(number):
    return number % 2 == 0

print(is_even(10))
print(is_even(7))
````

---

## 🔹 3. Validation Functions

Validation functions check whether data follows rules. They can return `True`/`False` or a helpful message.

````python
# ▶️ Validation pattern
def is_valid_password(password):
    if len(password) < 8:
        return False
    if " " in password:
        return False
    return True

print(is_valid_password("python123"))
print(is_valid_password("short"))
````

---

## 🔹 4. Transformation Functions

Transformation functions take data and return a changed version of it.

````python
# ▶️ Transformation pattern
def clean_name(name):
    return name.strip().title()

print(clean_name("  alice smith  "))
print(clean_name("BOB JONES"))
````

---

## 🔹 5. Aggregation Functions

Aggregation functions combine many values into one result, such as a total, average, minimum, or maximum.

````python
# ▶️ Aggregation pattern
def average(numbers):
    if len(numbers) == 0:
        return 0
    return sum(numbers) / len(numbers)

print(average([10, 20, 30]))
print(average([]))
````

---

## 🔹 6. Functions That Build Lists

Some functions create and return a new list. This is useful for filtering or transforming data.

````python
# ▶️ Build and return a list
def get_passing_scores(scores):
    passing = []

    for score in scores:
        if score >= 60:
            passing.append(score)

    return passing

print(get_passing_scores([45, 80, 72, 50, 91]))
````

---

## 🔹 7. Helper Functions

A helper function is a small function used by another function. This keeps code organized and easier to read.

````python
# ▶️ Helper function pattern
def add_tax(price):
    return price * 1.08

def format_price(price):
    return f"${price:.2f}"

def final_price_label(price):
    total = add_tax(price)
    return format_price(total)

print(final_price_label(20))
````

---

## 🔹 8. Default Parameter Pattern

Default parameters let a function work even if some arguments are not provided.

````python
# ▶️ Default parameter pattern
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(greet("Alice"))
print(greet("Bob", "Good morning"))
````
