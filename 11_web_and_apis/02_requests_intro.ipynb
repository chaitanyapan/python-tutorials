# 📘 Lesson 33: APIs with `requests`

An API, or **Application Programming Interface**, lets programs communicate with other programs. Web APIs usually send and receive data using HTTP. In Python, the popular `requests` package makes working with APIs easier.

| Concept | Description |
|---------|-------------|
| API | A way for programs to exchange data |
| Endpoint | A specific API URL |
| `requests.get()` | Send a GET request |
| `.status_code` | The HTTP status code |
| `.json()` | Convert JSON response data to Python data |
| `params` | Query string values sent with a request |

---

> ⚠️ This lesson uses the external `requests` package. If it is not installed, run `pip install requests` in your terminal. API examples also need an internet connection.

## 🔹 1. What Is an API?

A web API gives your Python program a URL it can request. The API sends back data, often as JSON.

Example idea:
```text
Python program → API request → Server
Python program ← JSON response ← Server
```

````python
# ▶️ API idea as dictionaries
request = {"method": "GET", "url": "https://api.example.com/users"}
response = {"status_code": 200, "data": [{"name": "Alice"}]}

print(request)
print(response)
````

## 🔹 3. Making a GET Request

A `GET` request asks an API for data. We will use JSONPlaceholder, a free fake API for practice.

````python
# ▶️ Basic GET request
import requests

try:
    response = requests.get("https://jsonplaceholder.typicode.com/posts/1")
    print("Status code:", response.status_code)
    print(response.text[:100])
except requests.RequestException as error:
    print("Request failed:", error)
````

---

## 🔹 4. Reading JSON Data

Many APIs return JSON. Use `.json()` to convert JSON into Python dictionaries or lists.

````python
# ▶️ Convert response JSON to Python data
import requests

try:
    response = requests.get("https://jsonplaceholder.typicode.com/posts/1")
    data = response.json()

    print(data)
    print("Title:", data["title"])
except requests.RequestException as error:
    print("Request failed:", error)
````

---

## 🔹 5. Checking Status Codes

Always check whether the request worked before using the data.

````python
# ▶️ Check status before reading data
import requests

try:
    response = requests.get("https://jsonplaceholder.typicode.com/posts/1")

    if response.status_code == 200:
        data = response.json()
        print("Request worked!")
        print(data["title"])
    else:
        print(f"Request failed with status {response.status_code}")
except requests.RequestException as error:
    print("Request failed:", error)
````

---

## 🔹 6. Getting a List of Data

Some API endpoints return a list. You can loop through the results.

````python
# ▶️ Loop through API list data
import requests

try:
    response = requests.get("https://jsonplaceholder.typicode.com/users")
    users = response.json()

    for user in users[:3]:
        print(f"{user['name']} - {user['email']}")
except requests.RequestException as error:
    print("Request failed:", error)
````

---

## 🔹 7. Query Parameters

Query parameters add extra information to a URL. With `requests`, pass them using the `params` argument.

````python
# ▶️ Send query parameters
import requests

try:
    params = {"userId": 1}
    response = requests.get("https://jsonplaceholder.typicode.com/posts", params=params)
    posts = response.json()

    print("Final URL:", response.url)
    print("Number of posts:", len(posts))
    print("First title:", posts[0]["title"])
except requests.RequestException as error:
    print("Request failed:", error)
````

---

## 🔹 8. Handling Errors

Network requests can fail. Use `try` / `except` and timeouts to make your program safer.

````python
# ▶️ Safer request with timeout
import requests

try:
    response = requests.get("https://jsonplaceholder.typicode.com/posts/1", timeout=5)
    response.raise_for_status()
    print(response.json())
except requests.Timeout:
    print("The request timed out.")
except requests.HTTPError as error:
    print("HTTP error:", error)
except requests.RequestException as error:
    print("Request problem:", error)
````

---

## 🔹 9. Simple API Helper Function

A helper function can keep API code reusable and easier to read.

````python
# ▶️ API helper function
import requests

def get_json(url, params=None):
    try:
        response = requests.get(url, params=params, timeout=5)
        response.raise_for_status()
        return response.json()
    except requests.RequestException as error:
        print("API request failed:", error)
        return None

data = get_json("https://jsonplaceholder.typicode.com/posts/1")
if data:
    print(data["title"])
````

````python
# ✅ Solution - Exercise 1: Fetch One User
import requests

try:
    response = requests.get("https://jsonplaceholder.typicode.com/users/1", timeout=5)
    response.raise_for_status()
    user = response.json()
    print(user["name"])
    print(user["email"])
except requests.RequestException as error:
    print("Request failed:", error)
````
