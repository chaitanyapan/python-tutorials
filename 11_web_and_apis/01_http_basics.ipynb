# 📘 Lesson 32: HTTP Basics

HTTP stands for **HyperText Transfer Protocol**. It is the main way web browsers, apps, and APIs send information over the internet.

| Concept | Description |
|---------|-------------|
| Client | The program making a request, like a browser or Python script |
| Server | The computer or service sending a response |
| Request | A message asking for data or an action |
| Response | The server's reply |
| URL | The address of a web resource |
| Status code | A number describing what happened |

---

## 🔹 1. What Is HTTP?

HTTP is a request-response system. A client sends a request, and a server sends back a response.

Example flow:
1. You visit a website.
2. Your browser sends an HTTP request.
3. The website's server sends an HTTP response.
4. Your browser shows the page.

````python
# ▶️ HTTP idea as plain text
client = "Browser"
server = "Website Server"

print(f"{client} sends a request to {server}.")
print(f"{server} sends back a response.")
````

---

## 🔹 2. URLs

A URL is a web address. It tells the client where to send the request.

Example:
```text
https://example.com/products?page=2
```

Common parts:
- `https` → protocol
- `example.com` → domain
- `/products` → path
- `?page=2` → query string

````python
# ▶️ URL parts example
url = "https://example.com/products?page=2"

print("URL:", url)
print("Protocol: https")
print("Domain: example.com")
print("Path: /products")
print("Query: page=2")
````

---

## 🔹 3. HTTP Methods

HTTP methods describe what action the client wants.

| Method | Common Use |
|--------|------------|
| `GET` | Read or fetch data |
| `POST` | Create or send new data |
| `PUT` | Replace existing data |
| `PATCH` | Update part of existing data |
| `DELETE` | Delete data |

````python
# ▶️ Method examples
actions = {
    "GET": "Read data",
    "POST": "Create data",
    "PUT": "Replace data",
    "PATCH": "Update part of data",
    "DELETE": "Delete data"
}

for method, meaning in actions.items():
    print(f"{method}: {meaning}")
````

---

## 🔹 4. Status Codes

Status codes tell us what happened after a request.

| Code | Meaning |
|------|---------|
| `200` | OK — request worked |
| `201` | Created — new resource made |
| `400` | Bad Request — client sent invalid data |
| `401` | Unauthorized — login/auth needed |
| `404` | Not Found — resource does not exist |
| `500` | Server Error — server had a problem |

````python
# ▶️ Check status code category
status_code = 404

if status_code >= 200 and status_code < 300:
    print("Success")
elif status_code >= 400 and status_code < 500:
    print("Client error")
elif status_code >= 500:
    print("Server error")
````

---

## 🔹 5. Request Headers

Headers are extra information sent with a request or response. They are often stored as key-value pairs.

````python
# ▶️ Headers as a dictionary
headers = {
    "Accept": "application/json",
    "User-Agent": "Python Tutorial"
}

for key, value in headers.items():
    print(f"{key}: {value}")
````

---

## 🔹 6. Request Body

A request body contains data sent to the server. `POST`, `PUT`, and `PATCH` often include a body.

APIs commonly use JSON for request bodies.

````python
# ▶️ JSON-style request body
new_user = {
    "name": "Alice",
    "email": "alice@example.com"
}

print(new_user)
````

---

## 🔹 7. Response Data

A response often includes data from the server. In APIs, this data is commonly JSON.

````python
# ▶️ Example response data
response = {
    "status_code": 200,
    "data": {
        "id": 1,
        "name": "Alice"
    }
}

print(response["status_code"])
print(response["data"]["name"])
````

---

## 🔹 8. Simulating an HTTP Request

Before using real web requests, we can simulate a request with dictionaries and functions.

````python
# ▶️ Simulate an HTTP request
def send_request(method, url, body=None):
    print(f"Sending {method} request to {url}")

    if body:
        print("Body:", body)

    return {"status_code": 200, "message": "Success"}

response = send_request("GET", "https://example.com/users")
print(response)
````

---

## 🔹 9. Using `urllib` for a Real GET Request

Python includes `urllib`, which can make basic web requests without installing extra packages.

> ⚠️ This cell needs an internet connection. If it fails, the notebook can still continue.

````python
# ▶️ Basic GET request with urllib
from urllib.request import urlopen

try:
    with urlopen("https://example.com") as response:
        print("Status code:", response.status)
        print("Content type:", response.headers.get("Content-Type"))
except Exception as error:
    print("Request failed:", error)
````
