## 🔹 3. Create a Reusable API Helper

A helper function keeps our API code simple and reusable.

```python
# ▶️ API helper function
import requests

def get_json(endpoint, params=None):
    url = BASE_URL + endpoint

    try:
        response = requests.get(url, params=params, timeout=5)
        response.raise_for_status()
        return response.json()
    except requests.Timeout:
        print("Request timed out.")
        return None
    except requests.HTTPError as error:
        print("HTTP error:", error)
        return None
    except requests.RequestException as error:
        print("Request failed:", error)
        return None

test_data = get_json("/users/1")
print(test_data)
```

---

## 🔹 4. Fetch One User

Let's write a function that gets one user by ID.

```python
# ▶️ Fetch one user
def get_user(user_id):
    return get_json(f"/users/{user_id}")

user = get_user(1)

if user:
    print(user["name"])
    print(user["email"])
    print(user["address"]["city"])
```

---

## 🔹 5. Display a User Nicely

A display function keeps output clean and readable.

```python
# ▶️ Display user
def display_user(user):
    if not user:
        print("No user to display.")
        return

    print(f"Name:    {user['name']}")
    print(f"Email:   {user['email']}")
    print(f"Phone:   {user['phone']}")
    print(f"Company: {user['company']['name']}")
    print(f"City:    {user['address']['city']}")

display_user(get_user(1))
```

---

## 🔹 6. Fetch All Users

Some API endpoints return a list of dictionaries.

```python
# ▶️ Fetch all users
def get_all_users():
    return get_json("/users")

users = get_all_users()

if users:
    for user in users[:5]:
        print(f"{user['id']}. {user['name']} - {user['email']}")
```

---

## 🔹 7. Search Users by Name

We can search through the users returned by the API.

```python
# ▶️ Search users
def search_users_by_name(users, search_text):
    results = []

    for user in users:
        if search_text.lower() in user["name"].lower():
            results.append(user)

    return results

matches = search_users_by_name(users, "Leanne") if users else []

for user in matches:
    display_user(user)
    print("---")
```

---

## 🔹 8. Fetch Posts for a User

Use query parameters to ask the API for posts belonging to a specific user.

```python
# ▶️ Fetch posts by user
def get_posts_by_user(user_id):
    return get_json("/posts", params={"userId": user_id})

posts = get_posts_by_user(1)

if posts:
    for post in posts[:3]:
        print(f"Post #{post['id']}: {post['title']}")
```

---

## 🔹 9. Summarize a User

Now combine user details with their posts.

```python
# ▶️ User summary
def display_user_summary(user_id):
    user = get_user(user_id)
    posts = get_posts_by_user(user_id)

    if not user or posts is None:
        print("Could not load user summary.")
        return

    print(f"User: {user['name']}")
    print(f"Email: {user['email']}")
    print(f"Number of posts: {len(posts)}")
    print("Recent posts:")

    for post in posts[:3]:
        print(f"- {post['title']}")

display_user_summary(1)
```
