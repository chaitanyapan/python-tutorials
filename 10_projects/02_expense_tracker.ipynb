## 🔹 1. Project Goal

Our expense tracker will be able to:

1. Add expenses
2. Display all expenses
3. Calculate the total spent
4. Filter expenses by category
5. Find the largest expense
6. Save expenses to a JSON file
7. Load expenses from a JSON file

> 💡 We will build the project step by step.

---

## 🔹 2. Create One Expense

One expense can be stored as a dictionary with a description, category, and amount.

```python
# ▶️ One expense as a dictionary
expense = {
    "description": "Lunch",
    "category": "Food",
    "amount": 12.50
}

print(expense)
print(expense["amount"])
```

---

## 🔹 3. Store Many Expenses

An expense tracker can store many expense dictionaries inside a list.

```python
# ▶️ Expense list
expenses = [
    {"description": "Lunch", "category": "Food", "amount": 12.50},
    {"description": "Bus ticket", "category": "Transport", "amount": 2.75},
    {"description": "Notebook", "category": "School", "amount": 4.99}
]

for expense in expenses:
    print(expense["description"], expense["amount"])
```

---

## 🔹 4. Add an Expense

A function makes it easy to add new expenses in the same format every time.

```python
# ▶️ Add expense function
def add_expense(expenses, description, category, amount):
    expense = {
        "description": description,
        "category": category,
        "amount": amount
    }
    expenses.append(expense)
    print(f"Added expense: {description} - ${amount:.2f}")

expenses = []
add_expense(expenses, "Lunch", "Food", 12.50)
add_expense(expenses, "Bus ticket", "Transport", 2.75)
add_expense(expenses, "Notebook", "School", 4.99)

print(expenses)
```

---

## 🔹 5. Display Expenses

A display function prints expenses in a readable way.

```python
# ▶️ Display expenses
def display_expenses(expenses):
    if len(expenses) == 0:
        print("No expenses found.")
        return

    for index, expense in enumerate(expenses, start=1):
        print(f"Expense #{index}")
        print(f"Description: {expense['description']}")
        print(f"Category:    {expense['category']}")
        print(f"Amount:      ${expense['amount']:.2f}")
        print("---")

display_expenses(expenses)
```

---

## 🔹 6. Calculate Total Spent

Loop through all expenses and add each amount to a total.

```python
# ▶️ Total expenses
def calculate_total(expenses):
    total = 0

    for expense in expenses:
        total += expense["amount"]

    return total

total = calculate_total(expenses)
print(f"Total spent: ${total:.2f}")
```

---

## 🔹 7. Filter by Category

Filtering lets us view only expenses from one category.

```python
# ▶️ Filter expenses by category
def filter_by_category(expenses, category):
    results = []

    for expense in expenses:
        if expense["category"].lower() == category.lower():
            results.append(expense)

    return results

food_expenses = filter_by_category(expenses, "Food")
display_expenses(food_expenses)
```

---

## 🔹 8. Find the Largest Expense

Track the biggest amount while looping through the expenses.

```python
# ▶️ Find largest expense
def find_largest_expense(expenses):
    if len(expenses) == 0:
        return None

    largest = expenses[0]

    for expense in expenses:
        if expense["amount"] > largest["amount"]:
            largest = expense

    return largest

largest = find_largest_expense(expenses)
print(largest)
```

---

## 🔹 9. Save and Load Expenses

Use JSON to save the expense list so it can be loaded later.

```python
# ▶️ Save and load expenses
import json

def save_expenses(expenses, filename):
    with open(filename, "w") as file:
        json.dump(expenses, file, indent=4)
    print(f"Saved {len(expenses)} expenses.")

def load_expenses(filename):
    try:
        with open(filename, "r") as file:
            return json.load(file)
    except FileNotFoundError:
        return []

save_expenses(expenses, "expenses.json")
loaded_expenses = load_expenses("expenses.json")
display_expenses(loaded_expenses)
```

---

## 🔹 10. Complete Expense Tracker Demo

Now let's combine the functions into one small workflow.

```python
# ▶️ Complete demo
expenses = load_expenses("expenses.json")

add_expense(expenses, "Coffee", "Food", 3.50)
add_expense(expenses, "Movie ticket", "Entertainment", 14.00)

print("All expenses:")
display_expenses(expenses)

print(f"Total spent: ${calculate_total(expenses):.2f}")

print("Food expenses:")
display_expenses(filter_by_category(expenses, "Food"))

print("Largest expense:")
print(find_largest_expense(expenses))

save_expenses(expenses, "expenses.json")
```
