## 🔹 2. Create a Contact

A contact can be stored as a dictionary with keys like `name`, `phone`, and `email`.

```python
# ▶️ One contact as a dictionary
contact = {
    "name": "Alice Smith",
    "phone": "555-1234",
    "email": "alice@example.com"
}

print(contact)
print(contact["name"])
```

---

## 🔹 3. Store Many Contacts

A contact book can be a list of contact dictionaries.

```python
# ▶️ Contact book as a list
contacts = [
    {"name": "Alice Smith", "phone": "555-1234", "email": "alice@example.com"},
    {"name": "Bob Jones", "phone": "555-5678", "email": "bob@example.com"}
]

for contact in contacts:
    print(contact["name"])
```

---

## 🔹 4. Add a Contact

Let's write a function that adds a new contact to the contact book.

```python
# ▶️ Add contact function
def add_contact(contacts, name, phone, email):
    contact = {
        "name": name,
        "phone": phone,
        "email": email
    }
    contacts.append(contact)
    print(f"Added contact: {name}")

contacts = []
add_contact(contacts, "Alice Smith", "555-1234", "alice@example.com")
add_contact(contacts, "Bob Jones", "555-5678", "bob@example.com")

print(contacts)
```

---

## 🔹 5. Display All Contacts

A display function makes contacts easier to read.

```python
# ▶️ Display contacts
def display_contacts(contacts):
    if len(contacts) == 0:
        print("No contacts found.")
        return

    for index, contact in enumerate(contacts, start=1):
        print(f"Contact #{index}")
        print(f"Name:  {contact['name']}")
        print(f"Phone: {contact['phone']}")
        print(f"Email: {contact['email']}")
        print("---")

display_contacts(contacts)
```

---

## 🔹 6. Search Contacts

We can search by checking whether the search text appears in each contact's name.

```python
# ▶️ Search contacts
def search_contacts(contacts, search_name):
    results = []

    for contact in contacts:
        if search_name.lower() in contact["name"].lower():
            results.append(contact)

    return results

matches = search_contacts(contacts, "alice")
display_contacts(matches)
```

---

## 🔹 7. Delete a Contact

To delete a contact, build a new list that keeps every contact except the one being removed.

```python
# ▶️ Delete contact by exact name
def delete_contact(contacts, name):
    updated_contacts = []
    deleted = False

    for contact in contacts:
        if contact["name"].lower() == name.lower():
            deleted = True
        else:
            updated_contacts.append(contact)

    if deleted:
        print(f"Deleted contact: {name}")
    else:
        print(f"No contact named {name} found.")

    return updated_contacts

contacts = delete_contact(contacts, "Bob Jones")
display_contacts(contacts)
```

---

## 🔹 8. Save Contacts to a JSON File

JSON is a good format for saving lists and dictionaries.

```python
# ▶️ Save contacts
import json

def save_contacts(contacts, filename):
    with open(filename, "w") as file:
        json.dump(contacts, file, indent=4)
    print(f"Saved {len(contacts)} contacts to {filename}.")

save_contacts(contacts, "contacts.json")
```

---

## 🔹 9. Load Contacts from a JSON File

When loading from a file, handle `FileNotFoundError` so the program does not crash if the file does not exist yet.

```python
# ▶️ Load contacts
import json

def load_contacts(filename):
    try:
        with open(filename, "r") as file:
            return json.load(file)
    except FileNotFoundError:
        return []

loaded_contacts = load_contacts("contacts.json")
display_contacts(loaded_contacts)
```

---

## 🔹 10. Complete Contact Book Demo

Here is a complete mini demo using all functions together.

```python
# ▶️ Complete contact book demo
contacts = load_contacts("contacts.json")

add_contact(contacts, "Charlie Brown", "555-0000", "charlie@example.com")
add_contact(contacts, "Diana Prince", "555-1111", "diana@example.com")

print("All contacts:")
display_contacts(contacts)

print("Search results for 'Diana':")
display_contacts(search_contacts(contacts, "Diana"))

save_contacts(contacts, "contacts.json")
```
