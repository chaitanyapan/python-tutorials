# 📘 Lesson 25: Loop Patterns

Loop patterns are common ways programmers use loops to solve problems.
Instead of memorizing every solution, learn the **pattern** behind it.

| Pattern | Description |
|---------|-------------|
| Counting | Count how many items match a condition |
| Accumulating | Build a total, string, or list step by step |
| Searching | Look for a specific value |
| Filtering | Keep only values that match a condition |
| Nested loops | Loop inside another loop for grids or combinations |

---

## 🔹 1. Counting Pattern

Use a counter variable when you need to count how many times something happens.

````python
# ▶️ Count even numbers
numbers = [3, 8, 12, 5, 7, 10, 14]
even_count = 0

for number in numbers:
    if number % 2 == 0:
        even_count += 1

print(f"Even numbers: {even_count}")
````

---

## 🔹 2. Accumulating Totals

Use an accumulator variable when you want to build a total step by step.

````python
# ▶️ Add numbers manually with a loop
prices = [10.99, 5.50, 3.25, 8.00]
total = 0

for price in prices:
    total += price

print(f"Total: ${total:.2f}")
````

---

## 🔹 3. Building a String

Loops can build strings one piece at a time.

````python
# ▶️ Build a comma-separated string
words = ["Python", "is", "fun"]
sentence = ""

for word in words:
    sentence += word + " "

print(sentence.strip())
````

---

## 🔹 4. Building a List

Start with an empty list, then append items during the loop.

````python
# ▶️ Build a list of squares
squares = []

for number in range(1, 6):
    squares.append(number ** 2)

print(squares)
````

---

## 🔹 5. Searching Pattern

Use a loop to search for an item. Use `break` when you find it.

````python
# ▶️ Search for a name
names = ["Alice", "Bob", "Charlie", "Diana"]
target = "Charlie"
found = False

for name in names:
    if name == target:
        found = True
        break

if found:
    print(f"Found {target}!")
else:
    print(f"{target} was not found.")
````

---

## 🔹 6. Finding Minimum and Maximum

You can track the smallest or largest value while looping.

````python
# ▶️ Find the largest number manually
scores = [72, 95, 81, 66, 88]
highest = scores[0]

for score in scores:
    if score > highest:
        highest = score

print(f"Highest score: {highest}")
````

---

## 🔹 7. Filtering Pattern

Filtering means keeping only items that match a condition.

````python
# ▶️ Filter passing scores
scores = [45, 82, 77, 59, 91, 38]
passing_scores = []

for score in scores:
    if score >= 60:
        passing_scores.append(score)

print(passing_scores)
````

---

## 🔹 8. Transforming Pattern

Transforming means creating new values from existing values.

````python
# ▶️ Convert names to uppercase
names = ["alice", "bob", "charlie"]
uppercase_names = []

for name in names:
    uppercase_names.append(name.upper())

print(uppercase_names)
````

---

## 🔹 9. Nested Loop Pattern

Nested loops are useful for grids, tables, and combinations.

````python
# ▶️ Print coordinate pairs
for row in range(1, 4):
    for column in range(1, 4):
        print(f"({row}, {column})", end=" ")
    print()
````
