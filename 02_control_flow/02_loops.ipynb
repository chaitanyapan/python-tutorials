# 📘 Lesson 4: Loops — `for` & `while`

Loops let you **repeat a block of code** multiple times without writing it over and over.
Python has two types of loops:

| Loop | Best used when... |
|------|------------------|
| `for` | You know **how many times** to repeat, or looping over a collection |
| `while` | You repeat **until a condition becomes False** |

---

## 🔹 1. `for` Loop

```python
for variable in sequence:
    # code to repeat
```

The loop runs **once for each item** in the sequence.

````python
# ▶️ Loop over a list
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
````

````python
# ▶️ Loop over a string (character by character)
for letter in "Python":
    print(letter)
````

---

## 🔹 2. `range()` — Loop a Specific Number of Times

`range(start, stop, step)` generates a sequence of numbers.

| Usage | Result |
|-------|--------|
| `range(5)` | 0, 1, 2, 3, 4 |
| `range(2, 6)` | 2, 3, 4, 5 |
| `range(0, 10, 2)` | 0, 2, 4, 6, 8 |
| `range(10, 0, -1)` | 10, 9, 8, ..., 1 |

> ⚠️ `stop` is **exclusive** — `range(5)` gives 0 to 4, not 5!

````python
# ▶️ range() examples
print("range(5):")
for i in range(5):
    print(i, end=" ")   # end=" " prints on same line

print("\n\nrange(2, 6):")
for i in range(2, 6):
    print(i, end=" ")

print("\n\nrange(0, 10, 2) — even numbers:")
for i in range(0, 10, 2):
    print(i, end=" ")

print("\n\nrange(5, 0, -1) — countdown:")
for i in range(5, 0, -1):
    print(i, end=" ")
````

---

## 🔹 3. `enumerate()` — Loop with Index

When you need both the **index** and the **value** in a loop.

````python
# ▶️ enumerate()
languages = ["Python", "JavaScript", "Go", "Rust"]

for index, language in enumerate(languages):
    print(f"{index + 1}. {language}")
````

---

## 🔹 4. `while` Loop

```python
while condition:
    # repeats as long as condition is True
```

> ⚠️ Always make sure the condition eventually becomes `False` — otherwise you get an **infinite loop**!

````python
# ▶️ Basic while loop
count = 1

while count <= 5:
    print(f"Count: {count}")
    count += 1   # ⚠️ Don't forget this! Without it → infinite loop

print("Done!")
````

````python
# ▶️ while loop — countdown
countdown = 5

while countdown > 0:
    print(f"{countdown}...")
    countdown -= 1

print("🚀 Blast off!")
````

---

## 🔹 5. `break` & `continue`

| Keyword | Effect |
|---------|--------|
| `break` | **Exits** the loop immediately |
| `continue` | **Skips** the current iteration and moves to the next |

````python
# ▶️ break — stop when we find 'banana'
fruits = ["apple", "mango", "banana", "cherry", "grape"]

for fruit in fruits:
    if fruit == "banana":
        print(f"Found {fruit}! Stopping.")
        break
    print(fruit)
````

````python
# ▶️ continue — skip odd numbers, print only even
for i in range(1, 11):
    if i % 2 != 0:
        continue     # skip odd numbers
    print(i, end=" ")
````

---

## 🔹 6. Nested Loops

A loop inside another loop. The **inner loop completes fully** for each iteration of the outer loop.

````python
# ▶️ Nested loops — multiplication table
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} x {j} = {i * j}")
    print("---")   # separator after each row
````

# ✅ Solutions

> 💡 **Try the exercises yourself first before looking at the solutions!**

````python
# ✅ Solution - Exercise 1: Sum of Numbers
total = 0

for i in range(1, 101):
    total += i

print("Sum:", total)   # 5050
````

````python
# ✅ Solution - Exercise 2: FizzBuzz with Loop
for i in range(1, 31):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
````

````python
# ✅ Solution - Exercise 3: Guessing Game
secret = 7
guess = 1
attempts = 0

while guess != secret:
    print(f"Trying: {guess}")
    attempts += 1
    guess += 1

attempts += 1   # count the final correct guess
print(f"🎉 Found it in {attempts} attempts!")
````

````python
# ✅ Solution - Exercise 4: Star Pattern
rows = 5

for i in range(1, rows + 1):
    for j in range(i):
        print("*", end="")
    print()   # newline after each row

# Bonus — same thing in one line using string multiplication:
print("\n--- Bonus one-liner version ---")
for i in range(1, rows + 1):
    print("*" * i)
````
