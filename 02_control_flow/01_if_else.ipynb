# 📘 Lesson 3: Control Flow — if / elif / else

Control flow lets your program **make decisions** based on conditions.
Instead of running every line top to bottom, Python can **choose which block of code to run**.

---

## 🔹 1. Basic `if` Statement

```python
if condition:
    # runs only if condition is True
```

⚠️ **Indentation matters in Python!** The code inside an `if` block must be indented (4 spaces or 1 tab).

````python
# ▶️ Basic if statement
age = 20

if age >= 18:
    print("You are an adult.")   # This runs because 20 >= 18 is True

print("This always runs.")       # Outside the if block
````

---

## 🔹 2. `if / else`

```python
if condition:
    # runs if condition is True
else:
    # runs if condition is False
```

````python
# ▶️ if / else
temperature = 15

if temperature > 25:
    print("It's warm outside, wear a t-shirt!")
else:
    print("It's cold outside, wear a jacket!")
````

---

## 🔹 3. `if / elif / else`

Use `elif` (else if) to check **multiple conditions**.
Python checks them **top to bottom** and stops at the first `True` condition.

```python
if condition1:
    ...
elif condition2:
    ...
elif condition3:
    ...
else:
    ...  # fallback if none matched
```

````python
# ▶️ if / elif / else — Grade checker
score = 72

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")

# Try changing 'score' and re-running!
````

---

## 🔹 4. Nested `if` Statements

You can place an `if` inside another `if` to check **multiple layers** of conditions.

````python
# ▶️ Nested if — Ticket price calculator
age = 25
is_member = True

if age >= 18:
    if is_member:
        print("Adult member ticket: $10")
    else:
        print("Adult ticket: $20")
else:
    if is_member:
        print("Child member ticket: $5")
    else:
        print("Child ticket: $12")
````

---

## 🔹 5. Combining Conditions with `and` / `or`

Instead of nesting, you can combine conditions in one line using logical operators.

````python
# ▶️ Combined conditions — same ticket example, cleaner!
age = 25
is_member = True

if age >= 18 and is_member:
    print("Adult member ticket: $10")
elif age >= 18 and not is_member:
    print("Adult ticket: $20")
elif age < 18 and is_member:
    print("Child member ticket: $5")
else:
    print("Child ticket: $12")
````

---

## 🔹 6. One-liner: Ternary (Inline) `if`

Python lets you write simple `if/else` in a single line:
```python
result = value_if_true if condition else value_if_false
```

````python
# ▶️ Ternary / inline if
age = 20
status = "adult" if age >= 18 else "minor"
print(f"You are a {status}.")

# Another example
number = -5
label = "positive" if number > 0 else "non-positive"
print(f"{number} is {label}.")
````
