# 📘 Lesson 11: Error Handling with `try` / `except`

Errors happen when programs run into problems.
Python lets you **handle errors gracefully** so your program does not crash unexpectedly.

| Concept | Description |
|---------|-------------|
| Exception | An error that happens while the program is running |
| `try` | Code that might cause an error |
| `except` | Code that runs if an error happens |
| `else` | Code that runs if no error happens |
| `finally` | Code that always runs |

---

## 🔹 1. What Is an Exception?

An exception is an error that happens while your program is running.
For example, dividing by zero or trying to open a missing file can cause exceptions.

---

## 🔹 2. Basic `try` / `except`

Use `try` for code that might fail.
Use `except` to decide what should happen if it fails.

```python
try:
    # risky code
except:
    # runs if an error happens
```

````python
# ▶️ Basic try / except
try:
    result = 10 / 0
    print(result)
except:
    print("Something went wrong!")

print("Program continues...")
````

---

## 🔹 3. Catching Specific Exceptions

It is better to catch specific errors instead of catching everything.
This makes your code clearer and safer.

````python
# ▶️ Catch specific exceptions
try:
    number = int("abc")
    print(number)
except ValueError:
    print("That value cannot be converted to an integer.")
````

---

## 🔹 4. Multiple `except` Blocks

Different errors can be handled in different ways.

````python
# ▶️ Multiple except blocks
try:
    x = int("10")
    y = int("0")
    print(x / y)
except ValueError:
    print("Please enter valid numbers.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
````

---

## 🔹 5. Using `else`

The `else` block runs only if no exception happens.

````python
# ▶️ try / except / else
try:
    age = int("25")
except ValueError:
    print("Invalid age.")
else:
    print(f"Age saved: {age}")
````

---

## 🔹 6. Using `finally`

The `finally` block always runs — whether an error happens or not.
It is commonly used for cleanup tasks.

````python
# ▶️ try / except / finally
try:
    result = 20 / 4
    print("Result:", result)
except ZeroDivisionError:
    print("Cannot divide by zero.")
finally:
    print("This always runs.")
````

---

## 🔹 7. Getting the Error Message

You can store the exception in a variable using `as`.

````python
# ▶️ Print the actual error message
try:
    print(10 / 0)
except ZeroDivisionError as error:
    print("Error message:", error)
````

---

## 🔹 8. Raising Your Own Errors

Use `raise` when you want to create an error yourself.
This is useful when invalid data should not be accepted.

````python
# ▶️ Raise your own exception
def set_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative.")
    print(f"Age set to {age}.")

try:
    set_age(-5)
except ValueError as error:
    print("Invalid input:", error)
````

# ✅ Solutions

````python
# ✅ Solution - Exercise 1: Safe Division
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Cannot divide by zero"

print(safe_divide(10, 2))
print(safe_divide(10, 0))
````

````python
# ✅ Solution - Exercise 2: Convert to Integer Safely
def to_int(value):
    try:
        return int(value)
    except ValueError:
        return None

print(to_int("42"))
print(to_int("hello"))
print(to_int(3.14))
````

````python
# ✅ Solution - Exercise 3: Read File Safely
def read_file_safe(filename):
    try:
        with open(filename, "r") as file:
            return file.read()
    except FileNotFoundError:
        return "File not found"

print(read_file_safe("missing_file.txt"))
````

````python
# ✅ Solution - Exercise 4: Validate Password
def validate_password(password):
    if len(password) < 8:
        raise ValueError("Password must be at least 8 characters long.")
    return "Password accepted"

try:
    print(validate_password("abc"))
except ValueError as error:
    print("Invalid password:", error)

try:
    print(validate_password("python123"))
except ValueError as error:
    print("Invalid password:", error)
````
