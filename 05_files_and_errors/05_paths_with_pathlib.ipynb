# 📘 Lesson 28: Paths with `pathlib`

`pathlib` is Python's modern way to work with file and folder paths. It helps you build paths, check if files exist, create folders, and read or write files more clearly.

| Concept | Description |
|---------|-------------|
| `Path` | Represents a file or folder path |
| `.exists()` | Checks if a path exists |
| `.is_file()` | Checks if a path is a file |
| `.is_dir()` | Checks if a path is a folder |
| `/` operator | Joins path parts safely |
| `.read_text()` / `.write_text()` | Read and write text files |

---

## 🔹 1. Importing `Path`

The most common way to use `pathlib` is to import `Path`.

```python
# ▶️ Import Path
from pathlib import Path

current_folder = Path(".")
print(current_folder)
print(current_folder.resolve())
```

---

## 🔹 2. Creating Path Objects

A `Path` object can represent a file or folder. It does not have to exist yet.

```python
# ▶️ Create Path objects
from pathlib import Path

file_path = Path("notes.txt")
folder_path = Path("data")

print(file_path)
print(folder_path)
```

---

## 🔹 3. Joining Paths with `/`

Use `/` to join path parts. This is cleaner and safer than manually typing slashes.

```python
# ▶️ Join path parts
from pathlib import Path

folder = Path("data")
file_path = folder / "students.txt"

print(file_path)
```

---

## 🔹 4. Creating Folders

Use `.mkdir()` to create a folder.

> 💡 `exist_ok=True` prevents an error if the folder already exists.

```python
# ▶️ Create a folder
from pathlib import Path

data_folder = Path("data")
data_folder.mkdir(exist_ok=True)

print("Folder ready:", data_folder)
```

---

## 🔹 5. Writing and Reading Text Files

`pathlib` can write and read text files without using `open()` directly.

```python
# ▶️ Write and read text
from pathlib import Path

file_path = Path("data") / "message.txt"
file_path.write_text("Hello from pathlib!")

content = file_path.read_text()
print(content)
```

---

## 🔹 6. Checking Paths

Use `.exists()`, `.is_file()`, and `.is_dir()` to inspect paths.

```python
# ▶️ Check paths
from pathlib import Path

file_path = Path("data") / "message.txt"
folder_path = Path("data")

print(file_path.exists())
print(file_path.is_file())
print(folder_path.is_dir())
```

---

## 🔹 7. Path Parts

A `Path` object gives useful information about a file path.

```python
# ▶️ Inspect path parts
from pathlib import Path

path = Path("data") / "report.txt"

print("Name:", path.name)
print("Stem:", path.stem)
print("Suffix:", path.suffix)
print("Parent:", path.parent)
```

---

## 🔹 8. Listing Files in a Folder

Use `.iterdir()` to loop through files and folders inside a directory.

```python
# ▶️ List folder contents
from pathlib import Path

folder = Path("data")

for path in folder.iterdir():
    print(path)
```

---

## 🔹 9. Finding Files with `glob()`

Use `.glob()` to find files matching a pattern.

Examples:
- `*.txt` means all text files in this folder
- `**/*.txt` means all text files in this folder and subfolders

```python
# ▶️ Find text files
from pathlib import Path

folder = Path("data")

for path in folder.glob("*.txt"):
    print("Text file:", path)
```
