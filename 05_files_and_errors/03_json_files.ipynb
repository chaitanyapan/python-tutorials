# 📘 Lesson 21: JSON Files

JSON, short for **JavaScript Object Notation**, is a common format for storing and sharing data. Python can read and write JSON using the built-in `json` module.

| Concept | Description |
|---------|-------------|
| JSON | A text format for structured data |
| `json.dumps()` | Convert Python data to a JSON string |
| `json.loads()` | Convert a JSON string to Python data |
| `json.dump()` | Write Python data to a JSON file |
| `json.load()` | Read JSON data from a file |

---

## 🔹 1. JSON and Python Data

JSON looks very similar to Python dictionaries and lists. It is often used in config files, APIs, and saved data.

````python
# ▶️ Python data that can become JSON
person = {
    "name": "Alice",
    "age": 30,
    "is_student": False,
    "skills": ["Python", "SQL", "Git"]
}

print(person)
print(type(person))
````

---

## 🔹 2. Convert Python Data to a JSON String

Use `json.dumps()` to convert Python data into a JSON string.

````python
# ▶️ json.dumps()
import json

person = {"name": "Alice", "age": 30, "skills": ["Python", "SQL"]}
json_text = json.dumps(person)

print(json_text)
print(type(json_text))
````

---

## 🔹 3. Pretty JSON

Use `indent` to make JSON easier to read.

````python
# ▶️ Pretty JSON
import json

settings = {
    "theme": "dark",
    "font_size": 14,
    "autosave": True
}

print(json.dumps(settings, indent=4))
````

---

## 🔹 4. Convert a JSON String to Python Data

Use `json.loads()` to convert a JSON string back into Python data.

````python
# ▶️ json.loads()
import json

json_text = '{"name": "Bob", "age": 25, "city": "London"}'
data = json.loads(json_text)

print(data)
print(type(data))
print(data["name"])
````

---

## 🔹 5. Write JSON to a File

Use `json.dump()` with `open()` to save data into a `.json` file.

````python
# ▶️ Write JSON file
import json

profile = {
    "username": "alice123",
    "level": 5,
    "points": 1200
}

with open("profile.json", "w") as file:
    json.dump(profile, file, indent=4)

print("Saved profile.json")
````

---

## 🔹 6. Read JSON from a File

Use `json.load()` to read JSON data from a file into Python.

````python
# ▶️ Read JSON file
import json

with open("profile.json", "r") as file:
    loaded_profile = json.load(file)

print(loaded_profile)
print(loaded_profile["username"])
````
