# 📘 Lesson 22: CSV Files

CSV stands for **Comma-Separated Values**. A CSV file stores table-like data as plain text. Python can read and write CSV files using the built-in `csv` module.

| Concept | Description |
|---------|-------------|
| CSV | A simple text format for rows and columns |
| `csv.reader()` | Read CSV rows as lists |
| `csv.writer()` | Write rows to a CSV file |
| `csv.DictReader()` | Read CSV rows as dictionaries |
| `csv.DictWriter()` | Write dictionaries to a CSV file |

---

## 🔹 1. What Is a CSV File?

A CSV file is often used for spreadsheets, reports, exports, and simple datasets. Each line is a row, and values are usually separated by commas.

Example CSV text:
```csv
name,age,city
Alice,30,New York
Bob,25,London
```

````python
# ▶️ Create a simple CSV file
csv_text = "name,age,city\nAlice,30,New York\nBob,25,London\n"

with open("people.csv", "w", newline="") as file:
    file.write(csv_text)

print("people.csv created!")
````

---

## 🔹 2. Reading CSV Files with `csv.reader()`

Use `csv.reader()` to read each row as a list of values.

````python
# ▶️ Read rows as lists
import csv

with open("people.csv", "r", newline="") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
````

---

## 🔹 3. Skipping the Header Row

Many CSV files have a first row called a **header**. It contains the column names. You can use `next(reader)` to skip or store it.

````python
# ▶️ Skip the header row
import csv

with open("people.csv", "r", newline="") as file:
    reader = csv.reader(file)
    header = next(reader)

    print("Header:", header)

    for row in reader:
        name = row[0]
        age = row[1]
        city = row[2]
        print(f"{name} is {age} years old and lives in {city}.")
````

---

## 🔹 4. Writing CSV Files with `csv.writer()`

Use `csv.writer()` to write rows to a CSV file. Each row is usually a list.

> 💡 Use `newline=""` when opening CSV files to avoid extra blank lines on some systems.

````python
# ▶️ Write rows to a CSV file
import csv

products = [
    ["name", "price", "quantity"],
    ["Notebook", 5.99, 10],
    ["Pen", 1.50, 25],
    ["Backpack", 29.99, 3]
]

with open("products.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(products)

print("products.csv written!")
````

---

## 🔹 5. Reading CSV Files as Dictionaries

`csv.DictReader()` uses the header row as dictionary keys. This often makes CSV code easier to read.

````python
# ▶️ Read rows as dictionaries
import csv

with open("people.csv", "r", newline="") as file:
    reader = csv.DictReader(file)

    for row in reader:
        print(row)
        print(f"Name: {row['name']} | Age: {row['age']} | City: {row['city']}")
````

---

## 🔹 6. Writing Dictionaries with `csv.DictWriter()`

`csv.DictWriter()` writes dictionaries to a CSV file. You provide the column names using `fieldnames`.

````python
# ▶️ Write dictionaries to a CSV file
import csv

students = [
    {"name": "Alice", "grade": "A", "score": 92},
    {"name": "Bob", "grade": "B", "score": 85},
    {"name": "Charlie", "grade": "A", "score": 95}
]

with open("students.csv", "w", newline="") as file:
    fieldnames = ["name", "grade", "score"]
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(students)

print("students.csv written!")
````

---

## 🔹 7. Basic CSV Data Analysis

When reading numbers from CSV files, remember that CSV values are read as strings. Convert values with `int()` or `float()` when needed.

````python
# ▶️ Calculate an average score from a CSV file
import csv

scores = []

with open("students.csv", "r", newline="") as file:
    reader = csv.DictReader(file)

    for row in reader:
        scores.append(int(row["score"]))

average = sum(scores) / len(scores)
print(f"Average score: {round(average, 2)}")
````

````python
# ✅ Solution - Exercise 1: Movies CSV
import csv

movies = [
    ["title", "year", "rating"],
    ["Toy Story", 1995, 8.3],
    ["Finding Nemo", 2003, 8.2],
    ["The Incredibles", 2004, 8.0]
]

with open("movies.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(movies)

with open("movies.csv", "r", newline="") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
````

````python
# ✅ Solution - Exercise 3: Total Inventory Value
import csv

def total_inventory_value(filename):
    total = 0

    with open(filename, "r", newline="") as file:
        reader = csv.DictReader(file)

        for row in reader:
            price = float(row["price"])
            quantity = int(row["quantity"])
            total += price * quantity

    return total

print(total_inventory_value("products.csv"))
````
