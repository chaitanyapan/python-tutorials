# 📘 Lesson 10: File Handling

File handling lets Python **read from** and **write to** files on your computer.
This is useful for saving data, reading logs, working with text files, and building real programs.

| Task | Common Method |
|------|---------------|
| Read a file | `open(..., "r")` |
| Write a file | `open(..., "w")` |
| Add to a file | `open(..., "a")` |
| Safely close files | `with open(...) as file:` |

---

## 🔹 1. Opening Files with `with open()`

The recommended way to work with files is using `with open(...)`.
Python automatically closes the file when the block finishes.

```python
with open("filename.txt", "mode") as file:
    # work with file here
```

````python
# ▶️ Write a simple text file
with open("hello.txt", "w") as file:
    file.write("Hello, Python files!")

print("File created successfully!")
````

---

## 🔹 2. File Modes

| Mode | Meaning | Behavior |
|------|---------|----------|
| `"r"` | Read | File must already exist |
| `"w"` | Write | Creates file or overwrites existing file |
| `"a"` | Append | Adds content to the end of file |
| `"x"` | Create | Creates a new file, error if it exists |

> ⚠️ Be careful with `"w"` — it overwrites the whole file!

````python
# ▶️ Write mode overwrites the file
with open("notes.txt", "w") as file:
    file.write("Line 1: Learning Python files.\n")
    file.write("Line 2: Writing text is easy.\n")

print("notes.txt written!")
````

---

## 🔹 3. Reading an Entire File

Use `.read()` to read all content from a file as one string.

````python
# ▶️ Read the entire file
with open("notes.txt", "r") as file:
    content = file.read()

print(content)
````

---

## 🔹 4. Reading Line by Line

Looping through a file reads it one line at a time.
This is useful for large files.

````python
# ▶️ Read line by line
with open("notes.txt", "r") as file:
    for line in file:
        print(line.strip())   # strip removes extra newline characters
````

---

## 🔹 5. Appending to a File

Append mode `"a"` adds new content to the end of a file without deleting existing content.

````python
# ▶️ Append new lines
with open("notes.txt", "a") as file:
    file.write("Line 3: Appending keeps old content.\n")

with open("notes.txt", "r") as file:
    print(file.read())
````

---

## 🔹 6. Writing Lists of Lines

Use `.writelines()` to write multiple strings.
Remember to include `\n` if you want each item on a new line.

````python
# ▶️ Write multiple lines
tasks = ["Study Python\n", "Practice loops\n", "Build a project\n"]

with open("tasks.txt", "w") as file:
    file.writelines(tasks)

with open("tasks.txt", "r") as file:
    print(file.read())
````

---

## 🔹 7. Checking if a File Exists

Before reading a file, you may want to check if it exists.
Python's `pathlib` module is a modern way to work with file paths.

````python
# ▶️ Check if a file exists
from pathlib import Path

path = Path("notes.txt")

if path.exists():
    print("notes.txt exists!")
else:
    print("notes.txt was not found.")
````
