# 📘 Lesson 17: Class Methods & Static Methods

Python classes can have regular methods, class methods, and static methods. Each one is useful in different situations.

| Method Type | First Parameter | Common Use |
|-------------|-----------------|------------|
| Instance method | `self` | Work with one object |
| Class method | `cls` | Work with the class itself |
| Static method | none | Utility function kept inside a class |

---

## 🔹 1. Review: Instance Methods

An **instance method** works with a specific object. It uses `self` to access that object's attributes.

```python
# ▶️ Instance method
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def describe(self):
        print(f"{self.name} has grade {self.grade}.")

student = Student("Alice", "A")
student.describe()
```

---

## 🔹 2. Class Variables

A **class variable** belongs to the class and is shared by all objects.

```python
# ▶️ Class variable
class Employee:
    company = "TechCorp"

    def __init__(self, name):
        self.name = name

employee1 = Employee("Alice")
employee2 = Employee("Bob")

print(employee1.company)
print(employee2.company)
print(Employee.company)
```

---

## 🔹 3. Class Methods

A **class method** works with the class itself instead of one specific object.

Use `@classmethod`, and use `cls` as the first parameter.

```python
# ▶️ Class method
class Employee:
    company = "TechCorp"

    def __init__(self, name):
        self.name = name

    @classmethod
    def change_company(cls, new_company):
        cls.company = new_company

employee = Employee("Alice")
print(employee.company)

Employee.change_company("CodeWorks")
print(employee.company)
print(Employee.company)
```

---

## 🔹 4. Class Methods as Alternative Constructors

Class methods are often used to create objects from different input formats.

```python
# ▶️ Alternative constructor
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def from_string(cls, text):
        name, age = text.split(",")
        return cls(name, int(age))

person = Person.from_string("Alice,30")
print(person.name)
print(person.age)
```

---

## 🔹 5. Static Methods

A **static method** belongs inside a class, but it does not need `self` or `cls`.

Use `@staticmethod` for helper functions related to the class.

```python
# ▶️ Static method
class MathTools:
    @staticmethod
    def add(a, b):
        return a + b

    @staticmethod
    def is_even(number):
        return number % 2 == 0

print(MathTools.add(5, 3))
print(MathTools.is_even(10))
print(MathTools.is_even(7))
```

```python
# ✅ Solution - Exercise 1: School Class Variable
class Student:
    school = "Python Academy"

    def __init__(self, name):
        self.name = name

student1 = Student("Alice")
student2 = Student("Bob")
print(student1.school)
print(student2.school)
```

```python
# ✅ Solution - Exercise 2: Change App Name
class AppConfig:
    app_name = "My App"

    @classmethod
    def rename_app(cls, new_name):
        cls.app_name = new_name

print(AppConfig.app_name)
AppConfig.rename_app("Task Tracker")
print(AppConfig.app_name)
```

```python
# ✅ Solution - Exercise 3: Create Product from Text
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    @classmethod
    def from_text(cls, text):
        name, price = text.split(",")
        return cls(name, float(price))

product = Product.from_text("Pen,2.5")
print(product.name)
print(product.price)
```

```python
# ✅ Solution - Exercise 4: Validation Helper
class Validator:
    @staticmethod
    def is_positive(number):
        return number > 0

print(Validator.is_positive(10))
print(Validator.is_positive(-3))
```
