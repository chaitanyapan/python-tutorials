# 📘 Lesson 14: Inheritance

Inheritance lets one class reuse code from another class. This helps avoid repetition and makes related classes easier to organize.

| Concept | Description |
|---------|-------------|
| Parent class | The class being inherited from |
| Child class | The class that inherits from another class |
| Inheritance | Reusing attributes and methods from a parent class |
| Override | Replacing a parent method with a child version |
| `super()` | Calls code from the parent class |

---

## 🔹 1. Why Use Inheritance?

Inheritance is useful when multiple classes share common behavior.

For example, `Dog` and `Cat` are both animals, so they can share code from an `Animal` class.

```python
# ▶️ Parent and child classes
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")

class Dog(Animal):
    pass

dog = Dog("Buddy")
dog.eat()
```

---

## 🔹 2. Adding Child Methods

A child class can use parent methods and also define its own methods.

```python
# ▶️ Child class with its own method
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")

class Dog(Animal):
    def bark(self):
        print(f"{self.name} says woof!")

dog = Dog("Buddy")
dog.eat()
dog.bark()
```

---

## 🔹 3. Method Overriding

A child class can **override** a method from the parent class by defining a method with the same name.

```python
# ▶️ Overriding a method
class Animal:
    def speak(self):
        print("The animal makes a sound.")

class Cat(Animal):
    def speak(self):
        print("The cat says meow.")

animal = Animal()
cat = Cat()

animal.speak()
cat.speak()
```

---

## 🔹 4. Using `super()`

Use `super()` to call the parent class version of a method.
This is common when a child class needs extra attributes.

```python
# ▶️ super() example
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, grade):
        super().__init__(name)
        self.grade = grade

    def describe(self):
        print(f"{self.name} is in grade {self.grade}.")

student = Student("Alice", 10)
student.describe()
```

---

## 🔹 5. Multiple Child Classes

Many child classes can inherit from the same parent class.

```python
# ▶️ Multiple child classes
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def start(self):
        print(f"{self.brand} is starting.")

class Car(Vehicle):
    def drive(self):
        print(f"{self.brand} is driving.")

class Boat(Vehicle):
    def sail(self):
        print(f"{self.brand} is sailing.")

car = Car("Toyota")
boat = Boat("Yamaha")

car.start()
car.drive()
boat.start()
boat.sail()
```

```python
# ✅ Solution - Exercise 1: Animal Inheritance
class Animal:
    def __init__(self, name):
        self.name = name

    def sleep(self):
        print(f"{self.name} is sleeping.")

class Bird(Animal):
    def fly(self):
        print(f"{self.name} is flying.")

bird = Bird("Tweety")
bird.sleep()
bird.fly()
```

```python
# ✅ Solution - Exercise 2: Override a Method
class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

rectangle = Rectangle(5, 3)
print("Area:", rectangle.area())
```

```python
# ✅ Solution - Exercise 3: Use super()
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department

    def describe(self):
        print(f"Name: {self.name} | Salary: ${self.salary} | Department: {self.department}")

manager = Manager("Alice", 75000, "Engineering")
manager.describe()
```

```python
# ✅ Solution - Exercise 4: Vehicle Types
class Vehicle:
    def move(self):
        print("The vehicle moves.")

class Car(Vehicle):
    def move(self):
        print("The car drives on the road.")

class Bicycle(Vehicle):
    def move(self):
        print("The bicycle pedals along the path.")

car = Car()
bicycle = Bicycle()

car.move()
bicycle.move()
```
