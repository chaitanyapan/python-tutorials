# 📘 Lesson 13: Classes & Objects

Object-Oriented Programming, or **OOP**, helps you organize code by grouping related data and behavior together.

| Concept | Description |
|---------|-------------|
| Class | A blueprint for creating objects |
| Object | A specific item created from a class |
| Attribute | Data stored inside an object |
| Method | A function that belongs to an object |
| `self` | Refers to the current object |

---

## 🔹 1. Creating a Class

A **class** is like a blueprint. It describes what an object should have and what it can do.

```python
class ClassName:
    # class code here
```

````python
# ▶️ Creating a simple class
class Dog:
    pass

my_dog = Dog()
print(my_dog)
````

---

## 🔹 2. The `__init__` Method

The `__init__` method runs automatically when you create a new object.
It is commonly used to set starting values for attributes.

````python
# ▶️ Class with attributes
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

my_dog = Dog("Buddy", 3)

print(my_dog.name)
print(my_dog.age)
````

---

## 🔹 3. Methods

A **method** is a function inside a class. Methods describe what an object can do.

> 💡 Methods usually include `self` as the first parameter.

````python
# ▶️ Adding methods
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} says woof!")

    def describe(self):
        print(f"{self.name} is {self.age} years old.")

my_dog = Dog("Buddy", 3)
my_dog.bark()
my_dog.describe()
````

---

## 🔹 4. Creating Multiple Objects

You can create many objects from the same class. Each object has its own attribute values.

````python
# ▶️ Multiple objects
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def introduce(self):
        print(f"Hi, I'm {self.name} and my grade is {self.grade}.")

student1 = Student("Alice", "A")
student2 = Student("Bob", "B")

student1.introduce()
student2.introduce()
````

---

## 🔹 5. Updating Attributes

Attributes can be changed after an object is created.

````python
# ▶️ Updating object attributes
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def show_balance(self):
        print(f"{self.owner}'s balance: ${self.balance}")

account = BankAccount("Alice", 100)
account.show_balance()

account.balance = 150
account.show_balance()
````

---

## 🔹 6. Class Variables

A **class variable** is shared by all objects created from the class.

````python
# ▶️ Class variables
class Car:
    wheels = 4   # class variable

    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

car1 = Car("Toyota", "Corolla")
car2 = Car("Honda", "Civic")

print(car1.brand, car1.model, car1.wheels)
print(car2.brand, car2.model, car2.wheels)
````

````python
# ✅ Solution - Exercise 3: Simple Counter
class Counter:
    def __init__(self):
        self.count = 0

    def increment(self):
        self.count += 1

    def reset(self):
        self.count = 0

    def show(self):
        print(f"Count: {self.count}")

counter = Counter()
counter.increment()
counter.increment()
counter.show()
counter.reset()
counter.show()
````

````python
# ✅ Solution - Exercise 4: Bank Account
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount}")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount}")
        else:
            print("Not enough balance.")

    def show_balance(self):
        print(f"{self.owner}'s balance: ${self.balance}")

account = BankAccount("Alice", 100)
account.show_balance()
account.deposit(50)
account.withdraw(30)
account.withdraw(200)
account.show_balance()
````
