# 📘 Lesson 15: Encapsulation & Properties

Encapsulation means keeping an object's data and behavior together, while controlling how that data is accessed or changed.

| Concept | Description |
|---------|-------------|
| Public attribute | Can be accessed directly |
| Private-style attribute | Starts with `_` or `__` to signal internal use |
| Getter | A method that reads a value |
| Setter | A method that changes a value safely |
| Property | A clean way to use getter/setter logic like an attribute |

---

## 🔹 1. Public Attributes

By default, attributes in Python are public. That means you can read and change them directly.

```python
# ▶️ Public attributes
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

student = Student("Alice", 90)
print(student.name)
print(student.grade)

student.grade = 95
print(student.grade)
```

---

## 🔹 2. Private-Style Attributes

Python uses naming conventions to show that an attribute is meant for internal use.

- `_name` means: internal use by convention
- `__name` triggers name mangling, making direct access harder

> ⚠️ Python does not make attributes completely private like some languages. It relies on conventions.

```python
# ▶️ Private-style attributes
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self._balance = balance

    def show_balance(self):
        print(f"Balance: ${self._balance}")

account = BankAccount("Alice", 100)
account.show_balance()
```

---

## 🔹 3. Getters and Setters

A **getter** reads a value. A **setter** changes a value safely.

This lets us validate data before saving it.

```python
# ▶️ Getter and setter methods
class Product:
    def __init__(self, name, price):
        self.name = name
        self._price = price

    def get_price(self):
        return self._price

    def set_price(self, price):
        if price >= 0:
            self._price = price
        else:
            print("Price cannot be negative.")

product = Product("Notebook", 5)
print(product.get_price())
product.set_price(7)
print(product.get_price())
product.set_price(-3)
```

---

## 🔹 4. Properties

Properties let you use getter/setter logic with normal attribute syntax.

Use `@property` for the getter and `@name.setter` for the setter.

```python
# ▶️ Using properties
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    @property
    def price(self):
        return self._price

    @price.setter
    def price(self, value):
        if value >= 0:
            self._price = value
        else:
            print("Price cannot be negative.")

product = Product("Pen", 2)
print(product.price)
product.price = 4
print(product.price)
product.price = -1
```

---

## 🔹 5. Read-Only Properties

If you define a property without a setter, it becomes read-only.

```python
# ▶️ Read-only property
class Circle:
    def __init__(self, radius):
        self.radius = radius

    @property
    def area(self):
        return 3.14 * self.radius ** 2

circle = Circle(5)
print(circle.area)
# circle.area = 100  # ❌ This would cause an error because area has no setter
```

```python
# ✅ Solution - Exercise 1: Safe Age
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):
        if value >= 0:
            self._age = value
        else:
            print("Age cannot be negative.")

person = Person("Alice", 30)
print(person.age)
person.age = -5
print(person.age)
```

```python
# ✅ Solution - Exercise 2: Temperature Property
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    @property
    def fahrenheit(self):
        return (self.celsius * 9/5) + 32

temp = Temperature(25)
print(f"Celsius: {temp.celsius}")
print(f"Fahrenheit: {temp.fahrenheit}")
```

```python
# ✅ Solution - Exercise 3: Bank Account Validation
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    @property
    def balance(self):
        return self._balance

    @balance.setter
    def balance(self, value):
        if value >= 0:
            self._balance = value
        else:
            print("Balance cannot be negative.")

account = BankAccount("Bob", 100)
print(account.balance)
account.balance = -50
print(account.balance)
```
