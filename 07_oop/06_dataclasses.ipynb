# 📘 Lesson 18: Dataclasses

Dataclasses make it easier to create classes that mainly store data. They reduce repeated code like `__init__`, `__repr__`, and comparisons.

| Concept | Description |
|---------|-------------|
| `@dataclass` | Decorator that auto-generates common methods |
| Field | An attribute defined in a dataclass |
| Default value | A value used when no argument is provided |
| `field()` | Helper for advanced defaults and options |
| Frozen dataclass | A dataclass whose values cannot be changed |

---

## 🔹 1. Why Dataclasses?

Without dataclasses, classes that only store data can require lots of repeated setup code.

```python
# ▶️ Regular class
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def __repr__(self):
        return f"Student(name={self.name}, grade={self.grade})"

student = Student("Alice", "A")
print(student)
```

---

## 🔹 2. Creating a Dataclass

Use `@dataclass` above a class. Then list the fields with type hints.

```python
# ▶️ Basic dataclass
from dataclasses import dataclass

@dataclass
class Student:
    name: str
    grade: str

student = Student("Alice", "A")
print(student)
print(student.name)
print(student.grade)
```

---

## 🔹 3. Default Values

Dataclass fields can have default values, just like function parameters.

```python
# ▶️ Dataclass with defaults
from dataclasses import dataclass

@dataclass
class User:
    username: str
    is_active: bool = True
    role: str = "member"

user1 = User("alice")
user2 = User("bob", False, "admin")

print(user1)
print(user2)
```

---

## 🔹 4. Comparing Dataclass Objects

Dataclasses automatically compare objects by their field values.

```python
# ▶️ Automatic comparison
from dataclasses import dataclass

@dataclass
class Point:
    x: int
    y: int

point1 = Point(2, 3)
point2 = Point(2, 3)
point3 = Point(5, 8)

print(point1 == point2)
print(point1 == point3)
```

---

## 🔹 5. Using `field()` for Lists

For mutable defaults like lists, use `field(default_factory=list)`.

> ⚠️ Avoid using `items: list = []` because that list can be shared unexpectedly.

```python
# ▶️ field(default_factory=list)
from dataclasses import dataclass, field

@dataclass
class ShoppingCart:
    owner: str
    items: list = field(default_factory=list)

cart = ShoppingCart("Alice")
cart.items.append("apple")
cart.items.append("notebook")
print(cart)
```

---

## 🔹 6. Frozen Dataclasses

A frozen dataclass cannot be changed after it is created. This is useful for values that should stay constant.

```python
# ▶️ Frozen dataclass
from dataclasses import dataclass

@dataclass(frozen=True)
class Color:
    red: int
    green: int
    blue: int

color = Color(255, 100, 50)
print(color)
# color.red = 0  # ❌ This would cause an error
```
