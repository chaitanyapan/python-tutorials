# 📘 Lesson 16: Polymorphism

Polymorphism means **many forms**. In Python, different objects can use the same method name but behave differently.

| Concept | Description |
|---------|-------------|
| Polymorphism | Same interface, different behavior |
| Method overriding | Child classes define their own version of a method |
| Duck typing | If an object has the needed method, Python can use it |
| Common interface | Different classes share the same method names |

---

## 🔹 1. Same Method Name, Different Behavior

Different classes can define methods with the same name. Python calls the correct method based on the object.

```python
# ▶️ Polymorphism with same method names
class Dog:
    def speak(self):
        print("Woof!")

class Cat:
    def speak(self):
        print("Meow!")

animals = [Dog(), Cat()]

for animal in animals:
    animal.speak()
```

---

## 🔹 2. Polymorphism with Inheritance

A parent class can define a method, and child classes can override it with their own behavior.

```python
# ▶️ Method overriding
class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

shapes = [Rectangle(4, 5), Circle(3)]

for shape in shapes:
    print("Area:", shape.area())
```

---

## 🔹 3. Duck Typing

Python often cares more about what an object **can do** than what type it is.

> 💡 If it has the method we need, we can use it.

```python
# ▶️ Duck typing
class EmailSender:
    def send(self, message):
        print(f"Sending email: {message}")

class SmsSender:
    def send(self, message):
        print(f"Sending SMS: {message}")

def notify(sender, message):
    sender.send(message)

notify(EmailSender(), "Your order shipped.")
notify(SmsSender(), "Your code is 1234.")
```

---

## 🔹 4. Built-In Polymorphism

Many built-in Python functions work with different types.

```python
# ▶️ Built-in polymorphism
print(len("Python"))
print(len([1, 2, 3]))
print(len({"name": "Alice", "age": 30}))

print(3 + 4)
print("Hello, " + "World!")
```

```python
# ✅ Solution - Exercise 1: Animal Sounds
class Cow:
    def sound(self):
        print("Moo!")

class Duck:
    def sound(self):
        print("Quack!")

class Sheep:
    def sound(self):
        print("Baa!")

animals = [Cow(), Duck(), Sheep()]
for animal in animals:
    animal.sound()
```

```python
# ✅ Solution - Exercise 2: Payment Methods
class CreditCard:
    def pay(self, amount):
        print(f"Paid ${amount} by credit card.")

class PayPal:
    def pay(self, amount):
        print(f"Paid ${amount} using PayPal.")

class Cash:
    def pay(self, amount):
        print(f"Paid ${amount} in cash.")

payments = [CreditCard(), PayPal(), Cash()]
for payment in payments:
    payment.pay(25)
```

```python
# ✅ Solution - Exercise 3: Shape Areas
class Square:
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side ** 2

class Triangle:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height / 2

shapes = [Square(4), Triangle(6, 3)]
for shape in shapes:
    print("Area:", shape.area())
```
