# 📘 Lesson 29: Regular Expressions

Regular expressions, often called **regex**, are patterns used to search, match, and clean text. Python supports regular expressions with the built-in `re` module.

| Concept | Description |
|---------|-------------|
| `re.search()` | Find the first match anywhere in text |
| `re.findall()` | Find all matches |
| `re.sub()` | Replace matching text |
| `.` | Match any single character |
| `*`, `+`, `?` | Match repeated characters |
| `\d`, `\w`, `\s` | Match digits, word characters, whitespace |

---

## 🔹 1. Importing `re`

Python's regular expression tools are in the built-in `re` module.

````python
# ▶️ Import re
import re

text = "Python is fun"
match = re.search("fun", text)

print(match)
print(match.group())
````

---

## 🔹 2. Searching for Text

`re.search()` looks for the first match. If it finds one, it returns a match object. If not, it returns `None`.

````python
# ▶️ Search example
import re

message = "Contact us at support@example.com"

if re.search("support", message):
    print("Found support address.")
else:
    print("No support address found.")
````

---

## 🔹 3. Finding All Matches

`re.findall()` returns a list of all matches.

````python
# ▶️ Find all numbers
import re

text = "Order 123 costs $45 and order 456 costs $78"
numbers = re.findall(r"\d+", text)

print(numbers)
````

---

## 🔹 4. Common Regex Symbols

| Pattern | Meaning | Example |
|---------|---------|---------|
| `\d` | Any digit | `0` to `9` |
| `\w` | Word character | letters, numbers, `_` |
| `\s` | Whitespace | spaces, tabs, newlines |
| `.` | Any character except newline | `a`, `7`, `!` |
| `+` | One or more | `\d+` matches many digits |
| `*` | Zero or more | `a*` |
| `?` | Zero or one | `colou?r` |

````python
# ▶️ Common symbols
import re

text = "User alice_123 has 42 points"

print(re.findall(r"\d+", text))   # digits
print(re.findall(r"\w+", text))   # words
print(re.findall(r"\s", text))    # spaces
````

---

## 🔹 5. Matching Start and End

Use `^` to match the start of a string and ` to match the end.

````python
# ▶️ Start and end patterns
import re

filename = "report.txt"

print(bool(re.search(r"^report", filename)))
print(bool(re.search(r"\.txt$", filename)))
````

---

## 🔹 6. Character Sets

Square brackets match one character from a set. For example, `[aeiou]` matches one vowel.

````python
# ▶️ Character sets
import re

word = "education"
vowels = re.findall(r"[aeiou]", word)

print(vowels)
print(len(vowels))
````

---

## 🔹 7. Groups with Parentheses

Parentheses create groups. Groups help you extract specific parts of a match.

````python
# ▶️ Extract groups
import re

date = "2026-05-22"
match = re.search(r"(\d{4})-(\d{2})-(\d{2})", date)

if match:
    print("Year:", match.group(1))
    print("Month:", match.group(2))
    print("Day:", match.group(3))
````

---

## 🔹 8. Replacing Text with `re.sub()`

Use `re.sub()` to replace text matching a pattern.

````python
# ▶️ Replace sensitive numbers
import re

text = "My phone number is 555-123-4567"
hidden = re.sub(r"\d", "X", text)

print(hidden)
````

---

## 🔹 9. Simple Email Pattern

Regex is often used to check or extract emails. Real-world email validation can be complex, but simple patterns are useful for practice.

````python
# ▶️ Find emails
import re

text = "Email alice@example.com or bob@test.org for help."
emails = re.findall(r"\w+@\w+\.\w+", text)

print(emails)
````
